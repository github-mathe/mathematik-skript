# 📖 Theoretischer Hintergrund


## <a id="#hospital"></a> Regel von L'Hospital

Die Berechnung des Grenzwertes

$$
\lim_{x \to x_0} \frac{f(x)}{g(x)}
$$

kann zu Schwierigkeiten führen, wenn sowohl $f(x)$ als auch $g(x)$
für $x \to x_0$ gegen $0$ oder $\pm \infty$ streben.

---
### Voraussetzungen

Seien $f, g : \mathbb{R} \to \mathbb{R}$ differenzierbar im gelochten Intervall

$$
(a,b) \setminus \{x_0\}
$$

und es gelte:

$$
\lim_{x \to x_0} f(x) = 0
$$

$$
\lim_{x \to x_0} g(x) = 0
$$

$$
g'(x) \neq 0 \quad \forall x \in (a,b) \setminus \{x_0\}
$$

---
### Voraussetzungen

Seien $f, g : \mathbb{R} \to \mathbb{R}$ differenzierbar im gelochten Intervall

$$
(a,b) \setminus \{x_0\}
$$

und es gelte:

$$
\lim_{x \to x_0} f(x) = 0
$$

$$
\lim_{x \to x_0} g(x) = 0
$$

$$
g'(x) \neq 0 \quad \forall x \in (a,b) \setminus \{x_0\}
$$

---
### Wichtiger Hinweis

Die Regel darf nur angewendet werden bei den unbestimmten Formen:

$$
\frac{0}{0}
\quad \text{oder} \quad
\frac{\infty}{\infty}.
$$

## <a id="#asymptoten"></a> Definitionen von Asymptoten

### Senkrechte Asymptote

Eine Gerade $$x = x_0$$ heißt senkrechte Asymptote der Funktion $$f$$,
wenn mindestens einer der folgenden Grenzwerte gilt:

$$
\lim_{x \to x_0^-} f(x) = \pm \infty
\quad \text{oder} \quad
\lim_{x \to x_0^+} f(x) = \pm \infty
$$

Bedeutung:  
Die Funktionswerte wachsen betragsmäßig unbegrenzt,
wenn sich $$x$$ der Stelle $$x_0$$ nähert.

---

### Waagerechte Asymptote

Eine Gerade $$y = L$$ heißt waagerechte Asymptote,
wenn gilt:

$$
\lim_{x \to \infty} f(x) = L
\quad \text{oder} \quad
\lim_{x \to -\infty} f(x) = L
$$

Bedeutung:  
Die Funktion nähert sich für große $$|x|$$ einem festen Wert $$L$$ an.

---

### Schräge Asymptote

Eine Gerade $$y = mx + b$$ heißt schräge Asymptote,
wenn gilt:

$$
\lim_{x \to \pm\infty} \big( f(x) - (mx+b) \big) = 0
$$

Zur Berechnung von $$m$$ und $$b$$ (falls die Grenzwerte existieren):

$$
m = \lim_{x \to \infty} \frac{f(x)}{x}
$$

$$
b = \lim_{x \to \infty} \big( f(x) - m x \big)
$$

Hinweis:  
Bei gebrochen-rationalen Funktionen mit

$$
\deg(\text{Zähler}) = \deg(\text{Nenner}) + 1
$$

bestimmt man die schräge Asymptote praktisch über Polynomdivision.

**2)** Berechnen Sie mit Hilfe der Regel von L’Hospital den Grenzwert

### <a id="#aufgabe-42-loesung"></a> 📘 Lösung Aufgabe 4.2 


#### **Aufgabe 4.2.1 - Lösung**
<a id="aufgabe2.1-loesung"></a>


Wir berechnen den Grenzwert

$$
L=\lim_{x\to 0}\frac{\ln(e^x-x)}{\ln(\cos x)}.
$$

---

Prüfung der Grenzwertform

Zähler:
$$
\ln(e^x-x)\xrightarrow[x\to 0]{}\ln(1-0)=\ln(1)=0
$$

Nenner:
$$
\ln(\cos x)\xrightarrow[x\to 0]{}\ln(1)=0
$$

Damit liegt die unbestimmte Form
$$
\frac{0}{0}
$$
vor.

L'Hospital ist anwendbar.

---

1. Anwendung von L'Hospital

**Zähler** $u(x)=\ln(e^x-x)$

Kettenregel:
$$
u'(x)=\frac{(e^x-x)'}{e^x-x}=\frac{e^x-1}{e^x-x}
$$

**Nenner** $v(x)=\ln(\cos x)$

Kettenregel:
$$
v'(x)=\frac{(\cos x)'}{\cos x}=\frac{-\sin x}{\cos x}=-\tan x
$$

Also:
$$
L=\lim_{x\to 0}\frac{\frac{e^x-1}{e^x-x}}{-\tan x}
=
\lim_{x\to 0}-\frac{e^x-1}{(e^x-x)\tan x}.
$$

---

Erneute Prüfung

Für $x\to 0$ gilt:
- $e^x-1\to 0$
- $e^x-x\to 1$
- $\tan x\to 0$

Damit geht der Nenner $(e^x-x)\tan x \to 1\cdot 0=0$ und der Zähler $\to 0$.

Wieder Form $\frac{0}{0}$, also **L'Hospital nochmal**.

Setze:
$$
A(x)=e^x-1,\qquad B(x)=(e^x-x)\tan x.
$$
Dann:
$$
L=\lim_{x\to 0}-\frac{A(x)}{B(x)}.
$$

---

2. Anwendung von L'Hospital

Ableitungen:

$$
A'(x)=e^x
$$

Für $B(x)=(e^x-x)\tan x$ (Produktregel):
$$
B'(x)=(e^x-x)'\tan x + (e^x-x)(\tan x)'
$$
$$
B'(x)=(e^x-1)\tan x + (e^x-x)\frac{1}{\cos^2 x}.
$$

Grenzwert einsetzen $x\to 0$:

- $(e^x-1)\tan x \to 0\cdot 0=0$
- $(e^x-x)\frac{1}{\cos^2 x}\to 1\cdot 1=1$

Also:
$$
\lim_{x\to 0}B'(x)=1
\qquad\text{und}\qquad
\lim_{x\to 0}A'(x)=1.
$$

Damit:
$$
L=\lim_{x\to 0}-\frac{A'(x)}{B'(x)}=-\frac{1}{1}=-1.
$$

---

Ergebnis

$$
\boxed{
\lim_{x\to 0}\frac{\ln(e^x-x)}{\ln(\cos x)}=-1
}
$$




```python
import numpy as np
import matplotlib.pyplot as plt

# ---------- Plot für 1) um x=0 ----------
def f1_num(t):
    return np.log(np.exp(t) - t) / np.log(np.cos(t))

eps = 1e-4
xs_left  = np.linspace(-0.6, -eps, 2000)
xs_right = np.linspace(eps, 0.6, 2000)
xs = np.concatenate([xs_left, xs_right])

ys = f1_num(xs)

plt.figure()
plt.axvline(0)
plt.axhline(-1)  # erwarteter Grenzwert
plt.plot(xs, ys)
plt.ylim(-3, 1)  # nur damit man’s gut sieht
plt.title(r"1) $\ln(e^x-x)/\ln(\cos x)$ nahe $x=0$ (Grenzwert = -1)")
plt.xlabel("x")
plt.ylabel("f1(x)")
plt.show()
```


####  **Aufgabe 4.2.2 - Lösung** 
<a id="aufgabe2.2-loesung"></a>

$$\lim_{x\to 1}\frac{x^2-1}{x^x-1}$$

---

Schritt 1: Unbestimmte Form prüfen

Setze $x=1$ ein:

Zähler:
$$
x^2 - 1 \Big|_{x=1} = 1^2 - 1 = 0
$$

Nenner:
$$
x^x - 1 \Big|_{x=1} = 1^1 - 1 = 0
$$

Damit liegt die unbestimmte Form

$$
\frac{0}{0}
$$

vor → **Regel von L’Hospital ist anwendbar.**

---

Schritt 2: L’Hospital anwenden

Wir leiten Zähler und Nenner getrennt ab.

🔹 Zähler

$$
\frac{d}{dx}(x^2 - 1) = 2x
$$

---

🔹 Nenner

Wir bestimmen zunächst die Ableitung von $x^x$.

Dazu schreiben wir:

$$
x^x = e^{x\ln x}
$$

Nun ableiten:

$$
\frac{d}{dx}e^{x\ln x}
=
e^{x\ln x}\cdot \frac{d}{dx}(x\ln x)
$$

Innere Ableitung:

$$
\frac{d}{dx}(x\ln x)
=
\ln x + 1
$$

Damit:

$$
\frac{d}{dx}(x^x)
=
e^{x\ln x}(\ln x+1)
=
x^x(\ln x+1)
$$

---

Schritt 3: Neuen Grenzwert berechnen

Nach L’Hospital ergibt sich:

$$
\lim_{x\to 1}\frac{x^2-1}{x^x-1}
=
\lim_{x\to 1}\frac{2x}{x^x(\ln x+1)}
$$

Nun setzen wir $x=1$ ein:

$$
2x \to 2
$$

$$
x^x \to 1
$$

$$
\ln x + 1 \to 0 + 1 = 1
$$

Damit:

$$
\lim_{x\to 1}\frac{2x}{x^x(\ln x+1)}
=
\frac{2}{1\cdot 1}
=
2
$$

---

Ergebnis

$$
\boxed{
\lim_{x\to 1}\frac{x^2-1}{x^x-1}=2
}
$$



```python
import numpy as np
import matplotlib.pyplot as plt

def f2(t):
    return (t**2 - 1) / (np.power(t, t) - 1)

eps = 1e-5
xs_left  = np.linspace(0.8, 1-eps, 2000)
xs_right = np.linspace(1+eps, 1.2, 2000)

xs = np.concatenate([xs_left, xs_right])
ys = f2(xs)

plt.figure()
plt.plot(xs, ys)

# Grenzwert-Linien
plt.axvline(1)
plt.axhline(2)

plt.title(r"$\frac{x^2-1}{x^x-1}$ nahe $x=1$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid()

plt.ylim(0, 4)
plt.xlim(0.8, 1.2)

plt.show()
```

### <a id="#aufgabe-44-loesung"></a> 📘 Lösung Aufgabe 4.4 

Wir betrachten den Grenzwert

$$
\lim_{x\to 1}\frac{1}{x-1}.
$$

Ein zweiseitiger Grenzwert existiert genau dann, wenn die beiden einseitigen Grenzwerte
$$
\lim_{x\to 1^-}\frac{1}{x-1}
\quad\text{und}\quad
\lim_{x\to 1^+}\frac{1}{x-1}
$$
existieren **und gleich sind**.

---

#### Rechtsseitiger Grenzwert ($x\to 1^+$)

Für $x\to 1^+$ gilt: $x>1$, also ist

$$
x-1>0
$$

und gleichzeitig wird $x-1$ **sehr klein positiv** ($x-1\to 0^+$). Dann wird der Kehrwert sehr groß positiv:

$$
\lim_{x\to 1^+}\frac{1}{x-1}=+\infty.
$$

---

#### Linksseitiger Grenzwert ($x\to 1^-$)

Für $x\to 1^-$ gilt: $x<1$, also ist

$$
x-1<0
$$

und gleichzeitig wird $x-1$ **sehr klein negativ** ($x-1\to 0^-$). Dann wird der Kehrwert sehr groß negativ:

$$
\lim_{x\to 1^-}\frac{1}{x-1}=-\infty.
$$

---

#### Schlussfolgerung

Die beiden einseitigen Grenzwerte sind verschieden:

$$
\lim_{x\to 1^+}\frac{1}{x-1}=+\infty
\neq
-\infty=\lim_{x\to 1^-}\frac{1}{x-1}.
$$

Damit existiert der zweiseitige Grenzwert **nicht**:

$$
\boxed{\lim_{x\to 1}\frac{1}{x-1}\ \text{existiert nicht.}}
$$

Zusatz: Bei $x=1$ hat $\frac{1}{x-1}$ eine **Polstelle** (vertikale Asymptote).

---

### ii \(\displaystyle \lim_{x\to 0}\sin\!\left(\frac{1}{x}\right)\)

Wir betrachten den Grenzwert

$$
\lim_{x\to 0}\sin\!\left(\frac{1}{x}\right).
$$

Schritt 1: Was passiert mit \(\frac{1}{x}\), wenn \(x\to 0\)?
- Für \(x\to 0^+\) gilt: \(\frac{1}{x}\to +\infty\)
- Für \(x\to 0^-\) gilt: \(\frac{1}{x}\to -\infty\)

Das Argument der Sinusfunktion wird also **betragsmäßig unendlich groß**.

Schritt 2: Sinus ist periodisch und oszilliert
Die Sinusfunktion erfüllt:
- \(\sin(t)\in[-1,1]\) für alle \(t\)
- sie ist periodisch und nimmt Werte **immer wieder** an

Wenn \(\frac{1}{x}\) gegen \(\pm\infty\) läuft, dann „läuft“ \(\sin\left(\frac{1}{x}\right)\) **unendlich oft** durch Werte zwischen \(-1\) und \(1\).

Schritt 3: Zeige „kein Grenzwert“ mit zwei Folgen (Standardargument)
Wir konstruieren zwei Folgen \(x_n\to 0\), die zu **verschiedenen Funktionswerten** führen.

Folge 1: Funktionswert wird \(0\)
Wähle
$$
x_n = \frac{1}{n\pi}.
$$
Dann gilt \(x_n\to 0\) für \(n\to\infty\) und

$$
\sin\!\left(\frac{1}{x_n}\right)
=
\sin(n\pi)=0.
$$

Also:
$$
\sin\!\left(\frac{1}{x_n}\right)\to 0.
$$

Folge 2: Funktionswert wird \(1\)
Wähle
$$
y_n=\frac{1}{\frac{\pi}{2}+2\pi n}.
$$
Dann gilt \(y_n\to 0\) und

$$
\sin\!\left(\frac{1}{y_n}\right)

## 📝 Aufgabe 4.6  

**Titel:** Differentiation


Gegeben ist die Funktion

$$
g(x)=
\begin{cases}
bx^3+cx^2+d, & x\le 1,\\[4pt]
\ln(x), & x>1.
\end{cases}
$$

Gesucht sind $b,c,d\in\mathbb{R}$, sodass $g$ auf $\mathbb{R}$ **zweimal differenzierbar** ist.  
Außerdem: Ist $g$ für diese Wahl auch **dreimal differenzierbar**?

---

### <a id="#aufgabe-46-loesung"></a> 📘 Lösung Aufgabe 4.6 


Bedingungen an der Klebestelle $x=1$


Damit $g$ auf ganz $\mathbb{R}$ zweimal differenzierbar ist, müssen an der Stelle $x=1$ gelten:

1. **Stetigkeit**: $g(1^-)=g(1^+)$  
2. **Stetigkeit der 1. Ableitung**: $g'(1^-)=g'(1^+)$  
3. **Stetigkeit der 2. Ableitung**: $g''(1^-)=g''(1^+)$  

---

Stetigkeit von $g$ bei $x=1$

Links (Polynom):
$$
g(1^-)=b\cdot 1^3+c\cdot 1^2+d=b+c+d
$$

Rechts (Logarithmus):
$$
g(1^+)=\ln(1)=0
$$

Also:
$$
b+c+d=0 \quad (1)
$$

---

Stetigkeit von $g'$ bei $x=1$

Für $x\le 1$:
$$
g'(x)=(bx^3+cx^2+d)'=3bx^2+2cx
$$
also
$$
g'(1^-)=3b+2c
$$

Für $x>1$:
$$
g'(x)=(\ln x)'=\frac{1}{x}
$$
also
$$
g'(1^+)=1
$$

Damit:
$$
3b+2c=1 \quad (2)
$$

---

4) Stetigkeit von $g''$ bei $x=1$

Für $x\le 1$:
$$
g''(x)=(3bx^2+2cx)'=6bx+2c
$$
also
$$
g''(1^-)=6b+2c
$$

Für $x>1$:
$$
g''(x)=\left(\frac{1}{x}\right)'=-\frac{1}{x^2}
$$
also
$$
g''(1^+)=-1
$$

Damit:
$$
6b+2c=-1 \quad (3)
$$

---

Gleichungssystem lösen

Aus (2) und (3):

\[
\begin{aligned}
3b+2c&=1\\
6b+2c&=-1
\end{aligned}
\]

Subtrahiere die erste Gleichung von der zweiten:
$$
(6b+2c)-(3b+2c)=-1-1 \;\Rightarrow\; 3b=-2 \;\Rightarrow\; b=-\frac{2}{3}
$$

Setze in (2) ein:
$$
3\left(-\frac{2}{3}\right)+2c=1
\;\Rightarrow\;
-2+2c=1
\;\Rightarrow\;
2c=3
\;\Rightarrow\;
c=\frac{3}{2}
$$

Setze in (1) ein:
$$
-\frac{2}{3}+\frac{3}{2}+d=0
\;\Rightarrow\;
d=-\left(-\frac{2}{3}+\frac{3}{2}\right)
=-\frac{5}{6}
$$

**Ergebnis:**
$$
\boxed{b=-\frac{2}{3},\quad c=\frac{3}{2},\quad d=-\frac{5}{6}}
$$

---

Ist $g$ damit auch dreimal differenzierbar?

Für $x\le 1$ ist
$$
g''(x)=6bx+2c
$$
also
$$
g'''(x)=(6bx+2c)'=6b
$$
Mit $b=-\frac{2}{3}$:
$$
g'''(x)=6\left(-\frac{2}{3}\right)=-4
$$
Damit:
$$
g'''(1^-)= -4
$$

Für $x>1$ gilt:
$$
g''(x)=-\frac{1}{x^2}=-x^{-2}
$$
also
$$
g'''(x)=2x^{-3}=\frac{2}{x^3}
$$
Damit:
$$
g'''(1^+)=2
$$

Da
$$
g'''(1^-)\neq g'''(1^+)\quad (-4\neq 2),
$$
ist $g''$ bei $x=1$ nicht differenzierbar.

 **Also ist $g$ für diese Wahl nicht dreimal differenzierbar.**[Detaillierte Erklärung und Lösungsweg eintragen]

Gegeben:
$$
f(x)=\ln(1+x)
$$

---

1) Definitionsbereich $D$

Damit der Logarithmus definiert ist, muss gelten:
$$
1+x>0 \;\Longleftrightarrow\; x>-1
$$

**Definitionsbereich:**
$$
D=(-1,\infty)
$$

---

2) Achsenschnittpunkte / Nullstellen

Nullstelle (x-Achse)
$$
f(x)=0 \;\Longleftrightarrow\; \ln(1+x)=0
$$
Exponentieren:
$$
1+x=e^0=1 \;\Longleftrightarrow\; x=0
$$

**Nullstelle:**
$$
x_0=0 \quad\Rightarrow\quad (0,0)
$$

y-Achsenabschnitt
y-Achse bedeutet $x=0$:
$$
f(0)=\ln(1)=0
$$

**Achsenschnittpunkt:**
$$
(0,0)
$$

---

3) Grenzwerte / Verhalten am Rand des Definitionsbereichs

Für $x\to -1^+$
$$
\lim_{x\to -1^+}\ln(1+x)=\ln(0^+)=-\infty
$$

**Folgerung:** Es gibt eine **senkrechte Asymptote** bei
$$
x=-1
$$

Für $x\to \infty$
$$
\lim_{x\to\infty}\ln(1+x)=\infty
$$

**Folgerung:** Keine waagrechte/schräge Asymptote nach oben, die Funktion wächst langsam aber unbegrenzt.

---

4) Erste Ableitung $f'(x)$, Monotonie, Extremstellen

Ableitung
$$
f(x)=\ln(1+x)
$$
Kettenregel:
$$
f'(x)=\frac{1}{1+x}
$$

Monotonie
Für $x>-1$ gilt:
$$
1+x>0 \;\Rightarrow\; \frac{1}{1+x}>0
$$

**Folgerung:**
- $f'(x)>0$ für alle $x\in(-1,\infty)$  
→ **streng monoton steigend**
- Da $f'(x)\neq 0$ im gesamten Definitionsbereich, gibt es **keine Extremstellen**.

---

5) Zweite Ableitung $f''(x)$, Krümmung, Wendestellen

Zweite Ableitung
$$
f'(x)=\frac{1}{1+x}=(1+x)^{-1}
$$
Ableiten:
$$
f''(x)=- (1+x)^{-2}=-\frac{1}{(1+x)^2}
$$

Krümmung
Für $x>-1$ ist $(1+x)^2>0$, also:
$$
f''(x)=-\frac{1}{(1+x)^2}<0
$$

**Folgerung:**
- Der Graph ist überall **rechtsgekrümmt / konkav** (nach unten geöffnet).
- Da $f''(x)$ nie $0$ wird, gibt es **keine Wendestelle**.

---

6) Zusammenfassung der wichtigsten Ergebnisse

- **Definitionsbereich:** $$(-1,\infty)$$
- **Nullstelle / Achsenschnittpunkt:** $$(0,0)$$
- **Asymptote:** senkrecht bei $$x=-1$$
- **Monotonie:** streng steigend auf $$(-1,\infty)$$
- **Extremstellen:** keine
- **Krümmung:** überall konkav ($f''(x)<0$)
- **Wendestellen:** keine

---

7) Skizzenhinweise (qualitativ)

- Der Graph startet bei $x\to -1^+$ bei $-\infty$ (Asymptote).
- Er steigt streng monoton an.
- Er geht durch $(0,0)$.
- Er bleibt stets konkav (die Steigung nimmt ab).
- Für große $x$ wächst er weiter, aber immer langsamer.

```python
# ============================
# Interaktives Kurvendiskussions-Tool (SymPy + Matplotlib + ipywidgets)
# ============================
# ✅ Funktionen:
# - f(x), f'(x), f''(x)
# - Definitionsbereich (reell & stetig, soweit bestimmbar)
# - Nullstellen, kritische Stellen, Wendestellen (soweit lösbar)
# - Asymptoten (heuristisch: vertikal / horizontal / schief bei rationalen Funktionen)
# - Plot inkl. Markierungen
#
# Voraussetzung (im Kernel installiert):
#   pip install sympy numpy matplotlib ipywidgets
# Hinweis: In VS Code müssen Widgets ggf. extra unterstützt werden (Jupyter Extension).
# Workaround ohne Widgets: unten "kurvendiskussion(...)" direkt aufrufen.

import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

from sympy.calculus.util import continuous_domain
from ipywidgets import interact, Text, FloatText, IntSlider, Checkbox

# Symbol
x = sp.Symbol("x", real=True)

# Erlaubte SymPy-Funktionen/Bezeichner für die Eingabe
_LOCALS = {
    "x": x,
    "ln": sp.log,
    "log": sp.log,
    "sqrt": sp.sqrt,
    "sin": sp.sin,
    "cos": sp.cos,
    "tan": sp.tan,
    "exp": sp.exp,
    "abs": sp.Abs,
    "pi": sp.pi,
}

def _try_real_finiteset_solutions(expr):
    """
    Versucht reelle Lösungen von expr=0 zu bestimmen.
    - Wenn FiniteSet -> Liste numerischer Werte
    - Sonst -> SymPy Set (Intervall/Union/ConditionSet/etc.)
    """
    try:
        sol = sp.solveset(sp.Eq(expr, 0), x, domain=sp.S.Reals)
        if isinstance(sol, sp.FiniteSet):
            vals = []
            for s in sol:
                sN = sp.N(s)
                if sN.is_real:
                    vals.append(float(sN))
            return sorted(set(vals))
        return sol
    except Exception:
        return None

def _asymptotes_heuristic(expr_simpl):
    """
    Heuristik:
    - Vertikal: Nullstellen des Nenners (bei rationalen Funktionen)
    - Horizontal: Grenzwerte x->±∞ (falls endlich)
    - Schief: Polynomdivision (bei rationalen Funktionen) und Quotient ist linear
    """
    asym = {"vertical": [], "horizontal": [], "oblique": []}

    # Vertikale Asymptoten: denom = 0 (nur zuverlässig für rationale Funktionen)
    try:
        num, den = sp.fraction(expr_simpl)
        if den != 1:
            den_roots = sp.solveset(sp.Eq(den, 0), x, domain=sp.S.Reals)
            if isinstance(den_roots, sp.FiniteSet):
                asym["vertical"] = sorted({float(sp.N(r)) for r in den_roots if sp.N(r).is_real})
            else:
                asym["vertical"] = den_roots
    except Exception:
        pass

    # Horizontale Asymptoten über Grenzwerte
    for side, target in [("+oo", sp.oo), ("-oo", -sp.oo)]:
        try:
            L = sp.limit(expr_simpl, x, target)
            if getattr(L, "is_real", False) and getattr(L, "is_finite", False):
                asym["horizontal"].append((side, sp.simplify(L)))
        except Exception:
            pass

    # Schiefe Asymptote über Polynomdivision (rational, Quotient linear)
    try:
        num, den = sp.fraction(expr_simpl)
        if den != 1:
            pn = sp.Poly(num, x)
            pd = sp.Poly(den, x)
            if pn.is_polynomial() and pd.is_polynomial():
                if pn.degree() >= pd.degree() + 1:
                    q, r = sp.div(pn, pd)
                    if q.degree() == 1:
                        asym["oblique"].append(sp.simplify(q.as_expr()))
    except Exception:
        pass

    return asym

def kurvendiskussion(expr_str, xmin=-5.0, xmax=5.0, n=2000, show_points=True, show_grid=True):
    # -------- Parse & simplify --------
    try:
        expr = sp.sympify(expr_str, locals=_LOCALS)
    except Exception as e:
        print("❌ Eingabe konnte nicht geparst werden. Prüfe SymPy-Syntax.")
        print("Fehler:", e)
        return

    expr_s = sp.simplify(expr)

    # -------- Domain (continuous real domain) --------
    try:
        dom = continuous_domain(expr_s, x, sp.S.Reals)
    except Exception:
        dom = "nicht sicher bestimmbar"

    # -------- Derivatives --------
    f1 = sp.simplify(sp.diff(expr_s, x))
    f2 = sp.simplify(sp.diff(f1, x))

    # -------- Special points (attempt) --------
    zeros = _try_real_finiteset_solutions(expr_s)
    crit  = _try_real_finiteset_solutions(f1)
    infl  = _try_real_finiteset_solutions(f2)

    # -------- Asymptotes (heuristic) --------
    asym = _asymptotes_heuristic(expr_s)

    # -------- Print analysis --------
    print("=== Kurvendiskussion ===")
    print("f(x)   =", expr_s)
    print("f'(x)  =", f1)
    print("f''(x) =", f2)
    print("\nDefinitionsbereich (reell & stetig):", dom)

    print("\nNullstellen:", zeros)
    print("Kritische Stellen (f'(x)=0):", crit)
    print("Wendestellen (f''(x)=0):", infl)

    print("\nAsymptoten (heuristisch):")
    print("  vertikal:", asym["vertical"])
    print("  horizontal:", asym["horizontal"])
    print("  schief:", asym["oblique"])

    # -------- Numeric plotting --------
    if xmax <= xmin:
        print("\n❌ xmin muss kleiner als xmax sein.")
        return

    f_np = sp.lambdify(x, expr_s, "numpy")

    xs = np.linspace(float(xmin), float(xmax), int(n))
    ys = np.full_like(xs, np.nan, dtype=float)

    # safe pointwise eval
    for i, xv in enumerate(xs):
        try:
            yv = f_np(xv)
            # skip complex/inf
            if np.isreal(yv) and np.isfinite(yv):
                ys[i] = float(np.real(yv))
        except Exception:
            pass

    plt.figure(figsize=(10, 5))
    plt.axhline(0)
    plt.axvline(0)
    if show_grid:
        plt.grid(True)

    plt.plot(xs, ys, label=f"f(x) = {sp.sstr(expr_s)}")

    # helper to mark points if we have finite list
    def _mark(points, label, marker):
        if not show_points:
            return
        if isinstance(points, list):
            first = True
            for p in points:
                if xmin < p < xmax:
                    try:
                        yp = float(sp.N(expr_s.subs(x, p)))
                        if np.isfinite(yp):
                            plt.plot([p], [yp], marker, markersize=8, label=label if first else None)
                            first = False
                    except Exception:
                        pass

    _mark(zeros if isinstance(zeros, list) else [], "Nullstelle", "o")
    _mark(crit  if isinstance(crit,  list) else [], "krit. Punkt", "s")
    _mark(infl  if isinstance(infl,  list) else [], "Wendepunkt", "D")

    # vertical asymptotes (numeric list only)
    if isinstance(asym["vertical"], list):
        for a in asym["vertical"]:
            if xmin < a < xmax:
                plt.axvline(a, linestyle="--")

    # horizontal asymptotes (finite)
    for side, L in asym["horizontal"]:
        try:
            yL = float(sp.N(L))
            if np.isfinite(yL):
                plt.axhline(yL, linestyle="--")
        except Exception:
            pass

    # oblique asymptotes (linear)
    for line in asym["oblique"]:
        try:
            line_np = sp.lambdify(x, line, "numpy")
            yl = line_np(xs)
            if np.all(np.isfinite(yl)):
                plt.plot(xs, yl, linestyle="--", label=f"schiefe Asymptote: {sp.sstr(line)}")
        except Exception:
            pass

    # robust ylim
    finite = ys[np.isfinite(ys)]
    if finite.size:
        pad = 0.1 * (np.nanmax(finite) - np.nanmin(finite) + 1e-9)
        plt.ylim(np.nanmin(finite) - pad, np.nanmax(finite) + pad)

    plt.legend()
    plt.show()


# -------- UI (Widgets) --------
def ui():
    """
    Interaktive Oberfläche:
    - f(x) Eingabe (SymPy Syntax)
    - xmin/xmax
    - Punkte (n)
    """
    return interact(
        kurvendiskussion,
        expr_str=Text(value="x*sqrt(16-x**2)", description="f(x) ="),
        xmin=FloatText(value=-5.0, description="xmin"),
        xmax=FloatText(value=5.0, description="xmax"),
        n=IntSlider(value=2000, min=500, max=8000, step=500, description="Punkte"),
        show_points=Checkbox(value=True, description="Punkte markieren"),
        show_grid=Checkbox(value=True, description="Gitter"),
    )

# Starte das Tool:
ui()

# Falls Widgets bei dir nicht rendern: nutze direkt
# kurvendiskussion("ln(1+x)", xmin=-0.9, xmax=5, n=3000)
```

#### 1)  $f(x) = \frac{x}{4 - x^2}$

```python
import numpy as np
import matplotlib.pyplot as plt

# Funktion definieren
def f(x):
    return x / (4 - x**2)

# x-Bereiche (ohne die Polstellen ±2)
x1 = np.linspace(-5, -2.1, 400)
x2 = np.linspace(-1.9, 1.9, 400)
x3 = np.linspace(2.1, 5, 400)

# Plot erstellen
plt.figure()

plt.plot(x1, f(x1))
plt.plot(x2, f(x2))
plt.plot(x3, f(x3))

# Vertikale Asymptoten
plt.axvline(-2, linestyle='--')
plt.axvline(2, linestyle='--')

# Waagerechte Asymptote
plt.axhline(0, linestyle='--')

plt.title("f(x) = x / (4 - x^2)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.ylim(-10, 10)
plt.grid(True)

plt.show()

# Kurz-Ergebnis ausgeben
print("Vertikale Asymptote: x = 2 und x = -2")
print("Horizontale Asymptote:   y = 0")
```


#### 2)  $h(x) = \frac{x^2 - 3x}{2x - 2}$

Restterm $\to 0$ für $x \to \pm\infty$ ⇒

$$
y = \frac{1}{2}x - 1
$$

---


```python
import numpy as np
import matplotlib.pyplot as plt

# Funktion: (x^2 - 3x) / (2x - 2)
def f(x):
    return (x**2 - 3*x) / (2*x - 2)

# --- Asymptoten bestimmen (kurz im Code dokumentiert) ---
# Vertikale Asymptote: 2x - 2 = 0 -> x = 1 (Zähler bei x=1 ist -2 != 0)
x_vert = 1

# Schiefe Asymptote durch Polynomdivision:
# (x^2 - 3x) / (2x - 2) = (1/2)x - 1 - 1/(x - 1)
# -> y = (1/2)x - 1
def asym_schraeg(x):
    return 0.5*x - 1

# Plot-Bereiche ohne Polstelle x=1
x1 = np.linspace(-5, 0.95, 600)
x2 = np.linspace(1.05, 5, 600)

plt.figure()

# Funktion plotten
plt.plot(x1, f(x1), label="f(x) = (x^2 - 3x)/(2x - 2)")
plt.plot(x2, f(x2))

# Vertikale Asymptote
plt.axvline(x_vert, linestyle="--", label="vertikal: x = 1")

# Schiefe Asymptote
xx = np.linspace(-5, 5, 400)
plt.plot(xx, asym_schraeg(xx), linestyle="--", label="schräg: y = 1/2 x - 1")

plt.title("Asymptoten von f(x) = (x^2 - 3x)/(2x - 2)")
plt.xlabel("x")
plt.ylabel("y")
plt.ylim(-10, 10)
plt.grid(True)
plt.legend()
plt.show()

# Kurz-Ergebnis ausgeben
print("Vertikale Asymptote: x = 1")
print("Schräge Asymptote:   y = (1/2) x - 1")
```

#### 3)  $f(x) = x^2 e^{-x}$

**Senkrechte Asymptoten:**

Der Nenner wird nie $0$ ⇒

Keine senkrechten Asymptoten.

**Waagerechte Asymptote:**

$$
\lim_{x \to +\infty} x^2 e^{-x}
=
\lim_{x \to +\infty} \frac{x^2}{e^x}
=
0
$$

⇒

$$
y = 0
$$

Für $x \to -\infty$ gilt $x^2 e^{-x} \to +\infty$ ⇒  
keine weitere Asymptote.

**Schräge Asymptote:**

Da der Grenzwert $0$ ist, existiert keine schräge Asymptote.



```python
import numpy as np
import matplotlib.pyplot as plt

# Funktion definieren
def f(x):
    return x**2 * np.exp(-x)

# x-Werte
x = np.linspace(-5, 10, 1000)

plt.figure()

plt.plot(x, f(x), label="f(x) = x^2 e^{-x}")

# Waagerechte Asymptote
plt.axhline(0, linestyle="--", label="y = 0")

plt.title("Asymptoten von f(x) = x^2 e^{-x}")
plt.xlabel("x")
plt.ylabel("y")
plt.ylim(-5, 50)
plt.grid(True)
plt.legend()

plt.show()

print("Waagerechte Asymptote für x → +∞: y = 0")
print("Keine vertikalen oder schrägen Asymptoten")
```


### <a id="#aufgabe4plot"></a> 📊 Interaktives Asymptoten-Tool

Dieses Tool visualisiert eine eingegebene Funktion \( f(x) \) zusammen mit ihren zugehörigen Asymptoten.

---

Eingabe

Gib eine Funktion in **SymPy-/Python-Syntax** ein, z.B.:

- `(x**2-1)/(x-1)`
- `1/(x-2)`
- `(2*x**2+3)/(5*x**2+7)`
- `x + 1/x`
- `sin(x)/x`

Verfügbare Funktionen:  
`sin`, `cos`, `tan`, `exp`, `log` (natürlicher Logarithmus), `sqrt`, `Abs`

Potenzschreibweise:  
`x**2` statt \( x^2 \)

---

🎨 Farbliche Kennzeichnung im Plot

| Farbe | Bedeutung |
|-------|------------|
| 🔵 Blau | Funktionsgraph \( f(x) \) |
| 🔴 Rot (gestrichelt) | Vertikale Asymptoten |
| 🟢 Grün (gestrichelt) | Horizontale Asymptoten |
| 🟠 Orange (gestrichelt) | Schräge Asymptoten |

---

🔍 Berechnungsmethoden

🔴 Vertikale Asymptoten

Eine vertikale Asymptote liegt vor, wenn

$$
\lim_{x \to a^\pm} f(x) = \pm \infty
$$

Das Tool untersucht Nullstellen des Nenners und prüft die einseitigen Grenzwerte.

---

🟢 Horizontale Asymptoten

Eine horizontale Asymptote liegt vor, wenn

$$
\lim_{x \to \pm\infty} f(x) = L
$$

mit einem endlichen Grenzwert \( L \).  
Dann gilt:

$$
y = L
$$

---

🟠 Schräge Asymptoten

Falls

$$
\lim_{x \to \pm\infty} \frac{f(x)}{x} = m \neq 0
$$

existiert, ergibt sich eine schräge Asymptote

$$
y = mx + b
$$

mit

$$
b = \lim_{x \to \pm\infty} \left(f(x) - mx\right)
$$

---

⚙️ Einstellmöglichkeiten

- $ x_{\min}, x_{\max} $ → sichtbarer Bereich der x-Achse  
- $ y_{\min}, y_{\max} $ → sichtbarer Bereich der y-Achse  
- Punktanzahl → Genauigkeit der Darstellung  

---

💡 Didaktischer Hinweis (Mathematik II)

Das Tool orientiert sich am klassischen Vorgehen der Kurvendiskussion:

1. Definitionslücken bestimmen  
2. Grenzwerte an Polstellen prüfen  
3. Grenzverhalten im Unendlichen untersuchen  
4. Gradvergleich / Polynomdivision für schräge Asymptoten  

Damit kannst du deine rechnerischen Ergebnisse visuell überprüfen.

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output

# =========================================================
# Interaktives Asymptoten-Tool (farbige Asymptoten + Legende)
# =========================================================

x = sp.Symbol("x", real=True)

def parse_expr(expr_str: str) -> sp.Expr:
    """
    Parsen einer Nutzereingabe in einen SymPy-Ausdruck.
    Erlaubt typische Funktionen und Konstanten.
    """
    locals_ = {
        "x": x,
        "pi": sp.pi,
        "E": sp.E,
        "sqrt": sp.sqrt,
        "abs": sp.Abs,
        "Abs": sp.Abs,
        "sin": sp.sin,
        "cos": sp.cos,
        "tan": sp.tan,
        "cot": sp.cot,
        "asin": sp.asin,
        "acos": sp.acos,
        "atan": sp.atan,
        "sinh": sp.sinh,
        "cosh": sp.cosh,
        "tanh": sp.tanh,
        "exp": sp.exp,
        "log": sp.log,  # log = ln
        "ln": sp.log,
        "sign": sp.sign,
    }
    return sp.sympify(expr_str, locals=locals_)

def real_unique(vals, tol=1e-7):
    vals = sorted(vals)
    out = []
    for v in vals:
        if not out or abs(v - out[-1]) > tol:
            out.append(v)
    return out

def safe_lambdify(expr: sp.Expr):
    """
    SymPy -> NumPy Funktion. Einige Ausdrücke liefern komplexe Werte;
    wir filtern später auf reelle/finite Werte.
    """
    return sp.lambdify(x, expr, modules=["numpy"])

# -------------------------
# Asymptoten-Berechnungen
# -------------------------

def vertical_asymptotes(expr: sp.Expr):
    """
    Kandidaten = reelle Nullstellen des Nenners.
    Vertikale Asymptote, wenn einseitiger Grenzwert ±∞ ist.
    Zusätzlich: hebbare Lücken (beide einseitigen Grenzwerte endlich).
    """
    expr_s = sp.together(expr)
    num, den = sp.fraction(expr_s)   # robust (statt sp.num/sp.denom)
    num = sp.factor(num)
    den = sp.factor(den)

    if den == 1:
        return [], [], []   # vertical, removable, undecided

    # Kandidaten: Lösungen den == 0
    sols = sp.solve(sp.Eq(den, 0), x)

    cand = []
    for s in sols:
        s = sp.simplify(s)
        try:
            sn = complex(sp.N(s))
            if abs(sn.imag) < 1e-10:
                cand.append(float(sn.real))
        except Exception:
            pass

    cand = real_unique(cand)

    vertical = []
    removable = []
    undecided = []

    for a in cand:
        try:
            Lp = sp.limit(expr_s, x, a, dir="+")
            Lm = sp.limit(expr_s, x, a, dir="-")

            infp = (Lp is sp.oo) or (Lp is -sp.oo)
            infm = (Lm is sp.oo) or (Lm is -sp.oo)

            if infp or infm:
                vertical.append((a, Lm, Lp))
            else:
                # typische hebbare Lücke: beidseitig endliches Limit
                if (getattr(Lp, "is_finite", False) is True) and (getattr(Lm, "is_finite", False) is True):
                    removable.append((a, Lm, Lp))
                else:
                    undecided.append((a, Lm, Lp))
        except Exception:
            undecided.append((a, None, None))

    return vertical, removable, undecided

def horizontal_asymptotes(expr: sp.Expr):
    """
    Horizontale Asymptoten: lim_{x->±∞} f(x) = L (endlich)
    """
    out = []
    for side in (sp.oo, -sp.oo):
        try:
            L = sp.limit(expr, x, side)
            if getattr(L, "is_finite", False) is True:
                out.append((side, sp.simplify(L)))
        except Exception:
            pass
    return out

def oblique_asymptotes(expr: sp.Expr):
    """
    Schräge Asymptote y = m x + b für x->±∞:
      m = lim f(x)/x (endlich, != 0)
      b = lim (f(x) - m x) (endlich)
    """
    out = []
    for side in (sp.oo, -sp.oo):
        try:
            m = sp.limit(expr / x, x, side)
            if getattr(m, "is_finite", False) is True and m != 0:
                b = sp.limit(expr - m*x, x, side)
                if getattr(b, "is_finite", False) is True:
                    out.append((side, sp.simplify(m), sp.simplify(b)))
        except Exception:
            pass
    return out

# -------------------------
# Plot + UI
# -------------------------

def plot_with_asymptotes(expr_str, x_min, x_max, y_min, y_max, n_points, show_removable):
    clear_output(wait=True)

    # 1) Parse
    try:
        expr = parse_expr(expr_str)
        expr_s = sp.simplify(expr)
    except Exception as e:
        print("❌ Konnte die Funktion nicht parsen.")
        print("Fehler:", e)
        print("\nBeispiele:")
        print("  1/(x-2)")
        print("  (x**2+1)/(x-1)")
        print("  (2*x**2+3)/(5*x**2+7)")
        print("  sin(x)/x")
        return

    # 2) Asymptoten
    vert, rem, undec = vertical_asymptotes(expr_s)
    horiz = horizontal_asymptotes(expr_s)
    obliq = oblique_asymptotes(expr_s)

    # 3) Textausgabe
    print("f(x) =", expr_s)

    if vert:
        print("\nVertikale Asymptoten (rot):")
        for a, Lm, Lp in vert:
            print(f"  x = {a:.6g}  (links: {Lm}, rechts: {Lp})")

    if horiz:
        print("\nHorizontale Asymptoten (grün):")
        for side, L in horiz:
            s = "+∞" if side == sp.oo else "-∞"
            print(f"  x → {s}:  y = {L}")

    if obliq:
        print("\nSchräge Asymptoten (orange):")
        for side, m, b in obliq:
            s = "+∞" if side == sp.oo else "-∞"
            print(f"  x → {s}:  y = ({m})*x + ({b})")

    if rem:
        print("\nHebbare Lücken (optional markiert):")
        for a, Lm, Lp in rem:
            # wenn beidseitig gleich, reicht ein Wert
            try:
                val = sp.simplify((Lm + Lp) / 2)
            except Exception:
                val = Lm
            print(f"  bei x = {a:.6g}  Grenzwert ≈ {val}")

    if undec:
        print("\n⚠️ Unklare Kandidaten (SymPy konnte nicht sauber entscheiden):")
        for a, Lm, Lp in undec:
            print(f"  x = {a:.6g}  (links: {Lm}, rechts: {Lp})")

    # 4) Plot
    f_np = safe_lambdify(expr_s)

    # Breakpoints: bei vertikalen Asymptoten segmentieren
    cuts = [float(x_min), float(x_max)]
    for a, _, _ in vert:
        if x_min < a < x_max:
            cuts.append(a)
    cuts = sorted(set(cuts))

    eps = 1e-3 * max(1.0, abs(x_max - x_min))

    plt.figure(figsize=(9, 5))

    # Funktionsgraph (blau) segmentiert
    first_fun = True
    for i in range(len(cuts) - 1):
        left, right = cuts[i], cuts[i + 1]

        # Abstand zu Polstellen
        if any(abs(right - a) < 1e-12 for a, _, _ in vert):
            right_plot = right - eps
        else:
            right_plot = right

        if any(abs(left - a) < 1e-12 for a, _, _ in vert):
            left_plot = left + eps
        else:
            left_plot = left

        if right_plot <= left_plot:
            continue

        xs = np.linspace(left_plot, right_plot, int(n_points))
        with np.errstate(all="ignore"):
            ys = f_np(xs)

        # nur reelle, finite Werte
        ys = np.real_if_close(ys, tol=1000)
        ys = np.where(np.isfinite(ys), ys, np.nan)

        plt.plot(
            xs, ys,
            linewidth=2,
            color="tab:blue",
            label="f(x)" if first_fun else None
        )
        first_fun = False

    # Achsen
    plt.axhline(0, linewidth=1, color="black")
    plt.axvline(0, linewidth=1, color="black")

    # Vertikale Asymptoten (rot)
    first_v = True
    for a, _, _ in vert:
        if x_min <= a <= x_max:
            plt.axvline(
                a, linestyle="--", linewidth=2,
                color="tab:red",
                label="vertikale Asymptote" if first_v else None
            )
            first_v = False

    # Horizontale Asymptoten (grün)
    first_h = True
    for side, L in horiz:
        try:
            Lf = float(sp.N(L))
            if y_min <= Lf <= y_max:
                plt.axhline(
                    Lf, linestyle="--", linewidth=2,
                    color="tab:green",
                    label="horizontale Asymptote" if first_h else None
                )
                first_h = False
        except Exception:
            pass

    # Schräge Asymptoten (orange)
    xs_line = np.linspace(x_min, x_max, 500)
    first_o = True
    for side, m, b in obliq:
        try:
            mf = float(sp.N(m))
            bf = float(sp.N(b))
            ys_line = mf * xs_line + bf
            plt.plot(
                xs_line, ys_line,
                linestyle="--", linewidth=2,
                color="tab:orange",
                label="schräge Asymptote" if first_o else None
            )
            first_o = False
        except Exception:
            pass

    # Hebbare Lücken markieren (offener Kreis)
    if show_removable and rem:
        first_r = True
        for a, Lm, Lp in rem:
            try:
                y_val = sp.limit(expr_s, x, a)
                y_val_f = float(sp.N(y_val))
            except Exception:
                continue

            if x_min <= a <= x_max and y_min <= y_val_f <= y_max:
                plt.plot(
                    [a], [y_val_f],
                    marker="o",
                    markersize=8,
                    markerfacecolor="white",
                    markeredgewidth=2,
                    markeredgecolor="black",
                    linestyle="None",
                    label="heb. Lücke" if first_r else None
                )
                first_r = False

    plt.xlim(x_min, x_max)
    plt.ylim(y_min, y_max)
    plt.grid(True, alpha=0.3)
    plt.title("Asymptoten-Tool: Funktion + farbige Asymptoten")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend(loc="best")
    plt.show()

# -------------------------
# Widgets
# -------------------------

func_input = widgets.Text(
    value="(x**2-1)/(x-1)",
    description="f(x)=",
    layout=widgets.Layout(width="600px")
)

x_min_w = widgets.FloatText(value=-10, description="x_min")
x_max_w = widgets.FloatText(value=10, description="x_max")
y_min_w = widgets.FloatText(value=-10, description="y_min")
y_max_w = widgets.FloatText(value=10, description="y_max")

n_points_w = widgets.IntSlider(value=1200, min=200, max=4000, step=100, description="Punkte")

show_removable_w = widgets.Checkbox(value=True, description="Hebbare Lücken markieren")

btn = widgets.Button(description="Plot aktualisieren", button_style="primary")

controls = widgets.VBox([
    func_input,
    widgets.HBox([x_min_w, x_max_w, y_min_w, y_max_w]),
    n_points_w,
    show_removable_w,
    btn
])

out = widgets.Output()

def on_click(_):
    with out:
        plot_with_asymptotes(
            expr_str=func_input.value,
            x_min=float(x_min_w.value),
            x_max=float(x_max_w.value),
            y_min=float(y_min_w.value),
            y_max=float(y_max_w.value),
            n_points=int(n_points_w.value),
            show_removable=bool(show_removable_w.value)
        )

btn.on_click(on_click)

display(controls, out)

# Initial plot
with out:
    plot_with_asymptotes(
        expr_str=func_input.value,
        x_min=float(x_min_w.value),
        x_max=float(x_max_w.value),
        y_min=float(y_min_w.value),
        y_max=float(y_max_w.value),
        n_points=int(n_points_w.value),
        show_removable=bool(show_removable_w.value)
    )
```
