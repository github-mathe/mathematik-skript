## 📚 Theoretische Grundlagen 

Diese Übersicht enthält die wichtigsten mathematischen Konzepte, die zur Bearbeitung der Aufgaben benötigt werden.

---

### 1. Grenzwerte (Limes)

Der Grenzwert beschreibt, welchem Wert sich eine Funktion nähert, wenn sich die Variable einem Punkt annähert.

Man schreibt

$$
\lim_{x \to x_0} f(x) = L
$$

wenn die Funktionswerte $f(x)$ beliebig nahe an $L$ herankommen, sobald $x$ ausreichend nahe bei $x_0$ liegt.

---

### 2. Einseitige Grenzwerte

Man unterscheidet Annäherung von links und von rechts.

Linksseitiger Grenzwert

$$
\lim_{x\to x_0^-} f(x)
$$

Rechtsseitiger Grenzwert

$$
\lim_{x\to x_0^+} f(x)
$$

Ein zweiseitiger Grenzwert existiert genau dann, wenn

$$
\lim_{x\to x_0^-} f(x)
=
\lim_{x\to x_0^+} f(x)
$$

gilt.

---

### 3. ε-Methode

Zur Berechnung einseitiger Grenzwerte kann man

rechtsseitig setzen

$$
x = x_0 + \varepsilon, \qquad \varepsilon \to 0^+
$$

linksseitig

$$
x = x_0 - \varepsilon, \qquad \varepsilon \to 0^+
$$

---

### 4. Wichtige Standardgrenzwerte

Trigonometrische Grenzwerte

$$
\lim_{x\to0}\frac{\sin x}{x}=1
$$

$$
\lim_{x\to0}\frac{\tan x}{x}=1
$$

Weitere wichtige Grenzwerte

$$
\lim_{x\to0^+} x\ln x = 0
$$

$$
\lim_{x\to\infty} e^{-x} = 0
$$

---

### 5. Regel von L’Hospital

Wenn ein Grenzwert die Form

$$
\frac{0}{0}
\quad \text{oder} \quad
\frac{\infty}{\infty}
$$

besitzt, darf man Zähler und Nenner getrennt ableiten.

$$
\lim_{x\to x_0}\frac{f(x)}{g(x)}
=
\lim_{x\to x_0}\frac{f'(x)}{g'(x)}
$$

Vorgehensweise:

1. Grenzwert einsetzen  
2. Unbestimmte Form prüfen  
3. Zähler und Nenner ableiten  
4. Grenzwert erneut berechnen  
5. Falls nötig Regel erneut anwenden  

---

### 6. Hyperbolische Funktionen

$$
\sinh x = \frac{e^x - e^{-x}}{2}
$$

$$
\cosh x = \frac{e^x + e^{-x}}{2}
$$

Diese Darstellung hilft besonders bei Grenzwerten für $x \to \infty$.

---

### 7. Wichtige Ableitungsregeln

Summenregel

$$
(f+g)' = f' + g'
$$

Produktregel

$$
(fg)' = f'g + fg'
$$

Quotientenregel

$$
\left(\frac{f}{g}\right)' =
\frac{f'g - fg'}{g^2}
$$

Kettenregel

$$
(f(g(x)))' = f'(g(x)) \cdot g'(x)
$$

---

### 8. Wichtige Ableitungen

$$
(e^x)' = e^x
$$

$$
(\ln x)' = \frac{1}{x}
$$

$$
(\sin x)' = \cos x
$$

$$
(\cos x)' = -\sin x
$$

$$
(x^n)' = nx^{n-1}
$$

---

### 9. Kurvendiskussion – Vorgehen

1. Definitionsbereich bestimmen  

2. Nullstellen berechnen  

$$
f(x)=0
$$

3. Ableitungen bestimmen  

$$
f'(x), \quad f''(x)
$$

4. Kritische Punkte bestimmen  

$$
f'(x)=0
$$

5. Art der Extremstelle bestimmen  

$$
f''(x)>0 \Rightarrow \text{Minimum}
$$

$$
f''(x)<0 \Rightarrow \text{Maximum}
$$

6. Wendepunkte bestimmen  

$$
f''(x)=0
$$

7. Monotonieverhalten mit $f'(x)$ untersuchen.

---

### 10. Asymptoten

Vertikale Asymptote

$$
\lim_{x\to x_0} f(x)=\pm\infty
\Rightarrow x=x_0
$$

Waagerechte Asymptote

$$
\lim_{x\to\pm\infty} f(x)=L
\Rightarrow y=L
$$

Schräge Asymptote

Wenn

$$
f(x) \sim ax+b
$$

für $x\to\infty$, dann ist

$$
y=ax+b
$$

die schräge Asymptote.

Bei rationalen Funktionen bestimmt man sie oft mit **Polynomdivision**.

```python
from IPython.display import display, Markdown
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets

# ============================
# EIN gemeinsamer Output
# ============================
main_out = widgets.Output()

# ============================
# Tipp
# ============================
tip_out = widgets.Output()
with tip_out:
    display(Markdown(r"""
<div style="padding: 15px;">

### 💡 Tipp

**Allgemein:**
- Ein zweiseitiger Grenzwert existiert nur, wenn **links = rechts** gilt:
$$
\lim_{x\to a^-} f(x) = \lim_{x\to a^+} f(x).
$$

**Zu a) i)**  
- Schau dir getrennt $x\to 1^-$ und $x\to 1^+$ an (Vorzeichen von $x-1$!).

**Zu a) ii)**  
- $\sin(\cdot)$ oszilliert zwischen $-1$ und $1$.
- Zeige „kein Grenzwert“ mit **zwei Folgen** $x_n\to 0$, die zu verschiedenen Funktionswerten führen.

**Zu b)**  
- Stückweise Funktion → **links** ($x<0$) und **rechts** ($x\ge 0$) getrennt einsetzen.

</div>
"""))

# ============================
# Lösung + Plots (erscheinen nur beim Aufklappen)
# ============================
sol_out = widgets.Output()

def plot_a_i():
    def f(x): return 1/(x-1)
    eps = 1e-4
    xs_left  = np.linspace(0.2, 1-eps, 2000)
    xs_right = np.linspace(1+eps, 1.8, 2000)

    plt.figure()
    plt.plot(xs_left, f(xs_left))
    plt.plot(xs_right, f(xs_right))
    plt.axvline(1)
    plt.title(r"$\frac{1}{x-1}$ nahe $x=1$ (links $\to -\infty$, rechts $\to +\infty$)")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.ylim(-20, 20)
    plt.grid()
    plt.show()

def plot_a_ii():
    def f(x): return np.sin(1/x)
    eps = 1e-4
    xs_left  = np.linspace(-0.2, -eps, 4000)
    xs_right = np.linspace(eps, 0.2, 4000)
    xs = np.concatenate([xs_left, xs_right])
    ys = f(xs)

    plt.figure()
    plt.plot(xs, ys)
    plt.axvline(0)
    plt.title(r"$\sin\!\left(\frac{1}{x}\right)$ nahe $x=0$ (oszilliert, kein Grenzwert)")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.ylim(-1.2, 1.2)
    plt.grid()
    plt.show()

def plot_b():
    def f(x):
        return np.where(x < 0, x**2, 3*x + 1)

    xs = np.linspace(-1, 1, 4000)
    ys = f(xs)

    plt.figure()
    plt.plot(xs, ys)
    plt.axvline(0)
    plt.title(r"Stückweise Funktion: links $x^2$, rechts $3x+1$ (Sprung bei $0$)")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.grid()
    plt.show()

with sol_out:
    display(Markdown(r"""
<div style="padding: 15px;">

## 📘 Lösung Aufgabe 4.4
<a id="aufgabe44-loesung"></a>

### a) i)  $\displaystyle \lim_{x\to 1}\frac{1}{x-1}$

Wir betrachten

$$
\lim_{x\to 1}\frac{1}{x-1}.
$$

Ein zweiseitiger Grenzwert existiert genau dann, wenn

$$
\lim_{x\to 1^-}\frac{1}{x-1}
\quad\text{und}\quad
\lim_{x\to 1^+}\frac{1}{x-1}
$$

existieren **und gleich sind**.

**Rechtsseitig** ($x\to 1^+$): $x-1\to 0^+$, daher

$$
\lim_{x\to 1^+}\frac{1}{x-1}=+\infty.
$$

**Linksseitig** ($x\to 1^-$): $x-1\to 0^-$, daher

$$
\lim_{x\to 1^-}\frac{1}{x-1}=-\infty.
$$

Da $+\infty\neq -\infty$, existiert der zweiseitige Grenzwert **nicht**:

$$
\boxed{\lim_{x\to 1}\frac{1}{x-1}\ \text{existiert nicht.}}
$$

---

### a) ii)  $\displaystyle \lim_{x\to 0}\sin\!\left(\frac{1}{x}\right)$

Wir betrachten

$$
\lim_{x\to 0}\sin\!\left(\frac{1}{x}\right).
$$

Da $\frac{1}{x}\to \pm\infty$ und $\sin$ periodisch ist, oszilliert
$\sin\!\left(\frac{1}{x}\right)$ unendlich oft zwischen $-1$ und $1$.

**Folgenargument:**

$$
x_n=\frac{1}{n\pi}\to 0
\quad\Rightarrow\quad
\sin\!\left(\frac{1}{x_n}\right)=\sin(n\pi)=0.
$$

$$
y_n=\frac{1}{\frac{\pi}{2}+2\pi n}\to 0
\quad\Rightarrow\quad
\sin\!\left(\frac{1}{y_n}\right)=\sin\!\left(\frac{\pi}{2}+2\pi n\right)=1.
$$

Damit gibt es keinen eindeutigen Grenzwert:

$$
\boxed{\lim_{x\to 0}\sin\!\left(\frac{1}{x}\right)\ \text{existiert nicht.}}
$$

---

### b)  $\displaystyle \lim_{x\to 0} f(x)$

$$
f(x)=
\begin{cases}
x^2, & x<0,\\[4pt]
3x+1, & x\ge 0.
\end{cases}
$$

**Linksseitig** ($x\to 0^-$):

$$
\lim_{x\to 0^-} f(x)=\lim_{x\to 0^-} x^2 = 0.
$$

**Rechtsseitig** ($x\to 0^+$):

$$
\lim_{x\to 0^+} f(x)=\lim_{x\to 0^+} (3x+1)=1.
$$

Da $0\neq 1$, existiert der zweiseitige Grenzwert **nicht**:

$$
\boxed{\lim_{x\to 0} f(x)\ \text{existiert nicht.}}
$$

---

## 📉 Grafische Verifikation

</div>
"""))

    # Plots (nur sichtbar beim Aufklappen)
    plot_a_i()
    plot_a_ii()
    plot_b()

# ============================
# Accordion (im weißen Kasten)
# ============================
acc = widgets.Accordion(children=[tip_out, sol_out])
acc.set_title(0, "💡 Tipp anzeigen")
acc.set_title(1, "📘 Lösung anzeigen (inkl. Plots)")

# ============================
# Alles in EINEM weißen Kasten
# ============================
with main_out:
    display(Markdown(r"""
<div style="
background-color: white;
padding: 25px;
border-radius: 16px;
box-shadow: 0 4px 16px rgba(0,0,0,0.08);
margin: 20px 0;
">

<a id="aufgabe4"></a>
# 📝 Aufgabe 4.4  
**Titel:** Funktionenlimes  

Berechnen Sie die folgenden Grenzwerte. Wenn der Grenzwert nicht existiert, begründen Sie Ihre Antwort.

### a)

**i)**  
$$
\lim_{x\to 1}\frac{1}{x-1}
$$

**ii)**  
$$
\lim_{x\to 0}\sin\!\left(\frac{1}{x}\right)
$$

---

### b)

Gegeben die stückweise definierte Funktion:
$$
f(x)=
\begin{cases}
x^2, & \text{falls } x<0,\\
3x+1, & \text{falls } x\ge 0.
\end{cases}
$$

Berechnen Sie
$$
\lim_{x\to 0} f(x).
$$

</div>
"""))
    display(acc)

display(main_out)
```

```python
from IPython.display import display, Markdown
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets

# ============================
# EIN gemeinsamer Output
# ============================
main_out = widgets.Output()

# ============================
# Tipp
# ============================
tip_out = widgets.Output()
with tip_out:
    display(Markdown(r"""
<div style="padding: 15px;">

### 💡 Tipp

Damit eine stückweise definierte Funktion bei der **Klebestelle** $x=1$ (hier der Übergang Polynom → $\ln x$)  
**zweimal differenzierbar** ist, müssen dort drei Dinge stimmen:

1. **Stetigkeit**: $g(1^-)=g(1^+)$  
2. **1. Ableitung stetig**: $g'(1^-)=g'(1^+)$  
3. **2. Ableitung stetig**: $g''(1^-)=g''(1^+)$  

👉 Das liefert ein Gleichungssystem für $b,c,d$.

**Für 3× differenzierbar** müsstest du zusätzlich prüfen:

4. **3. Ableitung stetig**: $g'''(1^-)=g'''(1^+)$

</div>
"""))

# ============================
# Lösung + Plot(s) (nur sichtbar beim Aufklappen)
# ============================
sol_out = widgets.Output()

def g_left(x, b, c, d):
    return b*x**3 + c*x**2 + d

def g_right(x):
    return np.log(x)

def plot_g(b, c, d):
    # Links und rechts getrennt, damit man die Klebestelle sauber sieht
    xs_left = np.linspace(-2, 1, 2000)
    xs_right = np.linspace(1 + 1e-4, 4, 2000)

    yl = g_left(xs_left, b, c, d)
    yr = g_right(xs_right)

    plt.figure()
    plt.plot(xs_left, yl)
    plt.plot(xs_right, yr)
    plt.axvline(1)
    plt.title(r"$g(x)$ mit $b=-\frac{2}{3},\,c=\frac{3}{2},\,d=-\frac{5}{6}$ (Klebestelle $x=1$)")
    plt.xlabel("x")
    plt.ylabel("g(x)")
    plt.grid()
    plt.show()

def plot_g_derivatives_at_1(b, c, d):
    # numerisch nah an 1 prüfen, ob 0., 1., 2. Ableitung "passt"
    # (rein zur Verifikation; mathematisch ist es über die Gleichungen bewiesen)
    eps = 1e-4

    xL = 1 - eps
    xR = 1 + eps

    # Funktionswerte
    gL = g_left(xL, b, c, d)
    gR = g_right(xR)

    # erste Ableitungen
    # links: g'(x)=3bx^2+2cx
    g1L = 3*b*xL**2 + 2*c*xL
    # rechts: g'(x)=1/x
    g1R = 1/xR

    # zweite Ableitungen
    # links: g''(x)=6bx+2c
    g2L = 6*b*xL + 2*c
    # rechts: g''(x)=-1/x^2
    g2R = -1/(xR**2)

    display(Markdown(rf"""
<div style="padding: 15px;">
<b>Numerische Kontrolle nahe $x=1$ (nur zur Plausibilität):</b>

- $g(1^-)\approx {gL:.6f}$,  $g(1^+)\approx {gR:.6f}$
- $g'(1^-)\approx {g1L:.6f}$, $g'(1^+)\approx {g1R:.6f}$
- $g''(1^-)\approx {g2L:.6f}$, $g''(1^+)\approx {g2R:.6f}$

(Hinweis: Wegen $\varepsilon$ sind das Näherungswerte.)
</div>
"""))

with sol_out:
    display(Markdown(r"""
<div style="padding: 15px;">

## 📘 Lösung Aufgabe 4.6
<a id="aufgabe46-loesung"></a>

### Bedingungen an der Klebestelle $x=1$

Damit $g$ auf ganz $\mathbb{R}$ **zweimal differenzierbar** ist, müssen an $x=1$ gelten:

1. **Stetigkeit**: $g(1^-)=g(1^+)$  
2. **Stetigkeit der 1. Ableitung**: $g'(1^-)=g'(1^+)$  
3. **Stetigkeit der 2. Ableitung**: $g''(1^-)=g''(1^+)$  

---

### 1) Stetigkeit von $g$ bei $x=1$

Links (Polynom):
$$
g(1^-)=b\cdot 1^3+c\cdot 1^2+d=b+c+d
$$

Rechts (Logarithmus):
$$
g(1^+)=\ln(1)=0
$$

Also:
$$
b+c+d=0 \quad (1)
$$

---

### 2) Stetigkeit von $g'$ bei $x=1$

Für $x\le 1$:
$$
g'(x)=(bx^3+cx^2+d)'=3bx^2+2cx
$$
$$
g'(1^-)=3b+2c
$$

Für $x>1$:
$$
g'(x)=(\ln x)'=\frac{1}{x}
$$
$$
g'(1^+)=1
$$

Damit:
$$
3b+2c=1 \quad (2)
$$

---

### 3) Stetigkeit von $g''$ bei $x=1$

Für $x\le 1$:
$$
g''(x)=(3bx^2+2cx)'=6bx+2c
$$
$$
g''(1^-)=6b+2c
$$

Für $x>1$:
$$
g''(x)=\left(\frac{1}{x}\right)'=-\frac{1}{x^2}
$$
$$
g''(1^+)=-1
$$

Damit:
$$
6b+2c=-1 \quad (3)
$$

---

### Gleichungssystem lösen

Aus (2) und (3):

$$
\begin{aligned}
3b+2c&=1\\
6b+2c&=-1
\end{aligned}
$$

Subtraktion:
$$
(6b+2c)-(3b+2c)=-1-1 \Rightarrow 3b=-2 \Rightarrow b=-\frac{2}{3}
$$

In (2) einsetzen:
$$
3\left(-\frac{2}{3}\right)+2c=1
\Rightarrow -2+2c=1
\Rightarrow 2c=3
\Rightarrow c=\frac{3}{2}
$$

In (1) einsetzen:
$$
-\frac{2}{3}+\frac{3}{2}+d=0
\Rightarrow d=-\left(-\frac{2}{3}+\frac{3}{2}\right)
=-\frac{5}{6}
$$

**Ergebnis:**
$$
\boxed{b=-\frac{2}{3},\quad c=\frac{3}{2},\quad d=-\frac{5}{6}}
$$

---

### Ist $g$ damit auch dreimal differenzierbar?

Für $x\le 1$:
$$
g''(x)=6bx+2c \quad\Rightarrow\quad g'''(x)=6b
$$
Mit $b=-\frac{2}{3}$:
$$
g'''(1^-)=6b=-4
$$

Für $x>1$:
$$
g''(x)=-\frac{1}{x^2} \Rightarrow g'''(x)=\frac{2}{x^3}
$$
$$
g'''(1^+)=2
$$

Da
$$
g'''(1^-)\neq g'''(1^+) \quad (-4\neq 2),
$$
ist $g$ **nicht** dreimal differenzierbar.

$$
\boxed{\text{$g$ ist zweimal, aber nicht dreimal differenzierbar.}}
$$

---

## 📉 Grafische Verifikation

</div>
"""))

    # Plots und Kontrolle (nur sichtbar im aufgeklappten Lösungsbereich)
    b, c, d = -2/3, 3/2, -5/6
    plot_g(b, c, d)
    plot_g_derivatives_at_1(b, c, d)

# ============================
# Accordion (im weißen Kasten)
# ============================
acc = widgets.Accordion(children=[tip_out, sol_out])
acc.set_title(0, "💡 Tipp anzeigen")
acc.set_title(1, "📘 Lösung anzeigen (inkl. Plot)")

# ============================
# Alles in EINEM weißen Kasten
# ============================
with main_out:
    display(Markdown(r"""
<div style="
background-color: white;
padding: 25px;
border-radius: 16px;
box-shadow: 0 4px 16px rgba(0,0,0,0.08);
margin: 20px 0;
">

<a id="aufgabe6"></a>
# 📝 Aufgabe 4.6  
**Titel:** Differentiation

Gegeben ist die Funktion

$$
g(x)=
\begin{cases}
bx^3+cx^2+d, & x\le 1,\\[4pt]
\ln(x), & x>1.
\end{cases}
$$

Gesucht sind $b,c,d\in\mathbb{R}$, sodass $g$ auf $\mathbb{R}$ **zweimal differenzierbar** ist.  
Außerdem: Ist $g$ für diese Wahl auch **dreimal differenzierbar**?

</div>
"""))
    display(acc)

display(main_out)
```

```python
from IPython.display import display, Markdown
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets

# ============================
# EIN gemeinsamer Output
# ============================
main_out = widgets.Output()

# ============================
# Tipp
# ============================
tip_out = widgets.Output()
with tip_out:
    display(Markdown(r"""
<div style="padding: 15px;">

### 💡 Tipp (Kurvendiskussion)

Gehe immer in dieser Reihenfolge vor:

1. **Definitionsbereich**: Bei $\ln(\,\cdot\,)$ muss das Argument **> 0** sein.  
2. **Achsenschnittpunkte / Nullstellen**: Setze $f(x)=0$ und/oder $x=0$.  
3. **Randverhalten**: Prüfe Grenzwerte an den Randstellen des Definitionsbereichs (hier besonders $x\to -1^+$).  
4. **1. Ableitung**: $f'(x)$ berechnen → Vorzeichen untersuchen → Monotonie/Extrema.  
5. **2. Ableitung**: $f''(x)$ berechnen → Vorzeichen untersuchen → Krümmung/Wendestellen.  
6. **Skizze**: Asymptote(n), wichtige Punkte, Steigen/Fallen, Krümmung eintragen.

</div>
"""))

# ============================
# Lösung + Plot(s) (nur sichtbar beim Aufklappen)
# ============================
sol_out = widgets.Output()

def plot_ln1p():
    # Plotbereich: nahe -1 plus weiter nach rechts
    xs_left = np.linspace(-0.99, -0.2, 1500)
    xs_right = np.linspace(-0.2, 5, 2500)
    xs = np.concatenate([xs_left, xs_right])

    ys = np.log(1 + xs)

    plt.figure()
    plt.plot(xs, ys)
    plt.axvline(-1)   # Asymptote
    plt.axhline(0)
    plt.axvline(0)
    plt.title(r"Skizze: $f(x)=\ln(1+x)$ (Asymptote bei $x=-1$)")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.grid()
    plt.show()

with sol_out:
    display(Markdown(r"""
<div style="padding: 15px;">

## 📘 Lösung Aufgabe 4.7
<a id="aufgabe47-loesung"></a>

Gegeben:
$$
f(x)=\ln(1+x)
$$

---

### 1) Definitionsbereich $D$

Damit der Logarithmus definiert ist:
$$
1+x>0 \;\Longleftrightarrow\; x>-1
$$

$$
\boxed{D=(-1,\infty)}
$$

---

### 2) Achsenschnittpunkte / Nullstellen

Nullstelle:
$$
\ln(1+x)=0 \Longleftrightarrow 1+x=1 \Longleftrightarrow x=0
$$

Damit:
$$
\boxed{(0,0)}
$$

y-Achse ($x=0$):
$$
f(0)=\ln(1)=0 \Rightarrow (0,0)
$$

---

### 3) Verhalten an Randstellen (Grenzwerte / Asymptote)

Für $x\to -1^+$:
$$
\lim_{x\to -1^+}\ln(1+x)=\ln(0^+)=-\infty
$$

→ **Senkrechte Asymptote**:
$$
\boxed{x=-1}
$$

Für $x\to \infty$:
$$
\lim_{x\to\infty}\ln(1+x)=\infty
$$

→ wächst unbegrenzt (langsam), keine waagrechte/schräge Asymptote.

---

### 4) Erste Ableitung $f'(x)$ (Monotonie, Extremstellen)

Kettenregel:
$$
f'(x)=\frac{1}{1+x}
$$

Für $x>-1$ gilt $1+x>0$, also:
$$
f'(x)>0 \quad\forall x\in(-1,\infty)
$$

→ **streng monoton steigend**.

Da $f'(x)$ nie $0$ wird:
$$
\boxed{\text{Keine Extremstellen.}}
$$

---

### 5) Zweite Ableitung $f''(x)$ (Krümmung, Wendestellen)

$$
f'(x)=(1+x)^{-1}
\Rightarrow
f''(x)=-(1+x)^{-2}=-\frac{1}{(1+x)^2}
$$

Für $x>-1$ ist $(1+x)^2>0$, daher:
$$
f''(x)<0 \quad\forall x\in(-1,\infty)
$$

→ überall **konkav** (rechtsgekrümmt / nach unten geöffnet).

Da $f''(x)$ nie $0$ wird:
$$
\boxed{\text{Keine Wendestellen.}}
$$

---

### 6) Zusammenfassung

- **Definitionsbereich:** $$(-1,\infty)$$  
- **Nullstelle / Achsenschnittpunkt:** $$(0,0)$$  
- **Asymptote:** $$x=-1$$  
- **Monotonie:** streng steigend auf $$(-1,\infty)$$  
- **Extremstellen:** keine  
- **Krümmung:** überall konkav ($f''(x)<0$)  
- **Wendestellen:** keine  

---

### 7) Skizze (qualitativ)

- Asymptote bei $x=-1$ mit $f(x)\to -\infty$ für $x\to -1^+$
- Graph geht durch $(0,0)$
- streng steigend, aber Steigung nimmt ab (konkav)
- wächst für $x\to\infty$ weiter, aber immer langsamer

---

## 📉 Plot / Skizze

</div>
"""))

    # Plot nur sichtbar im aufgeklappten Lösungsbereich
    plot_ln1p()

# ============================
# Accordion (im weißen Kasten)
# ============================
acc = widgets.Accordion(children=[tip_out, sol_out])
acc.set_title(0, "💡 Tipp anzeigen")
acc.set_title(1, "📘 Lösung anzeigen (inkl. Plot)")

# ============================
# Alles in EINEM weißen Kasten
# ============================
with main_out:
    display(Markdown(r"""
<div style="
background-color: white;
padding: 25px;
border-radius: 16px;
box-shadow: 0 4px 16px rgba(0,0,0,0.08);
margin: 20px 0;
">

<a id="aufgabe7"></a>
# 📝 Aufgabe 4.7  
**Titel:** Kurvendiskussion

Untersuchen Sie die folgende Funktion vollständig (Kurvendiskussion):

$$
f(x)=\ln(1+x)
$$

Bestimmen Sie dabei insbesondere:

- Definitionsbereich
- Nullstellen / Achsenschnittpunkte
- Verhalten an Randstellen des Definitionsbereichs (Grenzwerte, ggf. Asymptoten)
- Erste Ableitung $f'(x)$ (Monotonie, Extremstellen)
- Zweite Ableitung $f''(x)$ (Krümmung, Wendestellen)
- Skizze des Graphen (qualitativ)

</div>
"""))
    display(acc)

display(main_out)
```

```python
# ==========================================
# 📊 Interaktives Kurvendiskussions-Tool
# (schön formatiert im "weißen Kasten" Stil)
# ==========================================

from IPython.display import display, Markdown, clear_output
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
import ipywidgets as widgets
from sympy.calculus.util import continuous_domain

# -----------------------------
# Styling: ein gemeinsamer Output (kein "blauer Balken" zwischen Outputs)
# -----------------------------
main_out = widgets.Output()

# -----------------------------
# SymPy Setup
# -----------------------------
x = sp.Symbol("x", real=True)

_LOCALS = {
    "x": x,
    "ln": sp.log,
    "log": sp.log,
    "sqrt": sp.sqrt,
    "sin": sp.sin,
    "cos": sp.cos,
    "tan": sp.tan,
    "exp": sp.exp,
    "abs": sp.Abs,
    "Abs": sp.Abs,
    "pi": sp.pi,
}

def _try_real_finiteset_solutions(expr):
    """Versucht reelle Lösungen von expr=0 zu bestimmen; gibt Liste (FiniteSet) oder Set/None zurück."""
    try:
        sol = sp.solveset(sp.Eq(expr, 0), x, domain=sp.S.Reals)
        if isinstance(sol, sp.FiniteSet):
            vals = []
            for s in sol:
                sN = sp.N(s)
                if sN.is_real:
                    vals.append(float(sN))
            return sorted(set(vals))
        return sol
    except Exception:
        return None

def _asymptotes_heuristic(expr_simpl):
    """
    Heuristik:
    - Vertikal: Nullstellen des Nenners (bei rationalen Funktionen)
    - Horizontal: Grenzwerte x->±∞ (falls endlich)
    - Schief: Polynomdivision (bei rationalen Funktionen) und Quotient ist linear
    """
    asym = {"vertical": [], "horizontal": [], "oblique": []}

    # Vertikal: denom=0 (nur zuverlässig für rationale)
    try:
        num, den = sp.fraction(expr_simpl)
        if den != 1:
            den_roots = sp.solveset(sp.Eq(den, 0), x, domain=sp.S.Reals)
            if isinstance(den_roots, sp.FiniteSet):
                asym["vertical"] = sorted({float(sp.N(r)) for r in den_roots if sp.N(r).is_real})
            else:
                asym["vertical"] = den_roots
    except Exception:
        pass

    # Horizontal: Grenzwerte
    for side, target in [("+∞", sp.oo), ("-∞", -sp.oo)]:
        try:
            L = sp.limit(expr_simpl, x, target)
            if getattr(L, "is_real", False) and getattr(L, "is_finite", False):
                asym["horizontal"].append((side, sp.simplify(L)))
        except Exception:
            pass

    # Schief: Polynomdivision (Quotient linear)
    try:
        num, den = sp.fraction(expr_simpl)
        if den != 1:
            pn = sp.Poly(num, x)
            pd = sp.Poly(den, x)
            if pn.is_polynomial() and pd.is_polynomial():
                if pn.degree() >= pd.degree() + 1:
                    q, r = sp.div(pn, pd)
                    if q.degree() == 1:
                        asym["oblique"].append(sp.simplify(q.as_expr()))
    except Exception:
        pass

    return asym

def _format_set_or_list(obj):
    """Schöne Darstellung für Liste vs. SymPy-Set."""
    if isinstance(obj, list):
        if len(obj) == 0:
            return "keine (oder nicht gefunden)"
        return ", ".join([f"{v:.6g}" for v in obj])
    if obj is None:
        return "nicht sicher bestimmbar"
    return sp.sstr(obj)

# -----------------------------
# Kernfunktion: schreibt ALLES in ein Output-Panel (wird bei Slider/Inputs aktualisiert)
# -----------------------------
out_panel = widgets.Output()

def run_kurvendiskussion(expr_str, xmin, xmax, n, show_points, show_grid):
    with out_panel:
        clear_output(wait=True)

        # Parse
        try:
            expr = sp.sympify(expr_str, locals=_LOCALS)
        except Exception as e:
            display(Markdown(f"""
<div style="background:#fff3f3; padding:14px; border-radius:12px; border:1px solid #ffd1d1;">
<b>❌ Eingabe konnte nicht geparst werden.</b><br>
Prüfe die SymPy-Syntax (z.B. <code>sin(x)</code>, <code>ln(1+x)</code>, <code>sqrt(...)</code>, Potenzen mit <code>**</code>).<br><br>
<b>Fehler:</b> <code>{e}</code>
</div>
"""))
            return

        expr_s = sp.simplify(expr)

        # Domain
        try:
            dom = continuous_domain(expr_s, x, sp.S.Reals)
            dom_str = sp.sstr(dom)
        except Exception:
            dom_str = "nicht sicher bestimmbar"

        # Derivatives
        f1 = sp.simplify(sp.diff(expr_s, x))
        f2 = sp.simplify(sp.diff(f1, x))

        # Points
        zeros = _try_real_finiteset_solutions(expr_s)
        crit  = _try_real_finiteset_solutions(f1)
        infl  = _try_real_finiteset_solutions(f2)

        # Asymptotes
        asym = _asymptotes_heuristic(expr_s)

        # Header / Analysis block
        display(Markdown(rf"""
<div style="
background-color: white;
padding: 18px;
border-radius: 16px;
box-shadow: 0 4px 16px rgba(0,0,0,0.06);
margin: 10px 0;
">

### 📌 Analyse

**f(x)** = $$ {sp.latex(expr_s)} $$

**f'(x)** = $$ {sp.latex(f1)} $$

**f''(x)** = $$ {sp.latex(f2)} $$

**Definitionsbereich (reell & stetig, soweit bestimmbar):**  
$$ {sp.latex(dom) if dom_str != "nicht sicher bestimmbar" else r"\text{nicht sicher bestimmbar}"} $$

**Nullstellen:** {_format_set_or_list(zeros)}  
**Kritische Stellen (f'(x)=0):** {_format_set_or_list(crit)}  
**Wendestellen-Kandidaten (f''(x)=0):** {_format_set_or_list(infl)}

**Asymptoten (heuristisch):**
- vertikal: {_format_set_or_list(asym["vertical"])}
- horizontal: {", ".join([f"{side}: y={sp.sstr(L)}" for side, L in asym["horizontal"]]) if asym["horizontal"] else "keine (oder nicht gefunden)"}
- schief: {", ".join([sp.sstr(line) for line in asym["oblique"]]) if asym["oblique"] else "keine (oder nicht gefunden)"}

</div>
"""))

        # Numeric plotting
        if xmax <= xmin:
            display(Markdown("""
<div style="background:#fff3f3; padding:14px; border-radius:12px; border:1px solid #ffd1d1;">
<b>❌ Fehler:</b> <code>xmin</code> muss kleiner als <code>xmax</code> sein.
</div>
"""))
            return

        f_np = sp.lambdify(x, expr_s, "numpy")
        xs = np.linspace(float(xmin), float(xmax), int(n))
        ys = np.full_like(xs, np.nan, dtype=float)

        # safe evaluation
        for i, xv in enumerate(xs):
            try:
                yv = f_np(xv)
                if np.isreal(yv) and np.isfinite(yv):
                    ys[i] = float(np.real(yv))
            except Exception:
                pass

        plt.figure(figsize=(10, 5))
        plt.axhline(0)
        plt.axvline(0)
        if show_grid:
            plt.grid(True)

        plt.plot(xs, ys)

        # mark special points
        def _mark(points, marker):
            if not show_points:
                return
            if isinstance(points, list):
                for p in points:
                    if xmin < p < xmax:
                        try:
                            yp = float(sp.N(expr_s.subs(x, p)))
                            if np.isfinite(yp):
                                plt.plot([p], [yp], marker, markersize=8)
                        except Exception:
                            pass

        _mark(zeros if isinstance(zeros, list) else [], "o")
        _mark(crit  if isinstance(crit,  list) else [], "s")
        _mark(infl  if isinstance(infl,  list) else [], "D")

        # vertical asymptotes (numeric list)
        if isinstance(asym["vertical"], list):
            for a in asym["vertical"]:
                if xmin < a < xmax:
                    plt.axvline(a, linestyle="--")

        # horizontal asymptotes (finite)
        for side, L in asym["horizontal"]:
            try:
                yL = float(sp.N(L))
                if np.isfinite(yL):
                    plt.axhline(yL, linestyle="--")
            except Exception:
                pass

        # oblique asymptotes
        for line in asym["oblique"]:
            try:
                line_np = sp.lambdify(x, line, "numpy")
                yl = line_np(xs)
                if np.all(np.isfinite(yl)):
                    plt.plot(xs, yl, linestyle="--")
            except Exception:
                pass

        # robust y-limits
        finite = ys[np.isfinite(ys)]
        if finite.size:
            pad = 0.1 * (np.nanmax(finite) - np.nanmin(finite) + 1e-9)
            plt.ylim(np.nanmin(finite) - pad, np.nanmax(finite) + pad)

        plt.title(f"Plot: f(x) = {sp.sstr(expr_s)}")
        plt.xlabel("x")
        plt.ylabel("f(x)")
        plt.show()


# -----------------------------
# UI (schön & stabil): Controls + Output
# -----------------------------
expr_in = widgets.Text(value="x*sqrt(16-x**2)", description="f(x) =", layout=widgets.Layout(width="520px"))
xmin_in = widgets.FloatText(value=-5.0, description="xmin", layout=widgets.Layout(width="180px"))
xmax_in = widgets.FloatText(value=5.0,  description="xmax", layout=widgets.Layout(width="180px"))
n_in    = widgets.IntSlider(value=2000, min=500, max=8000, step=500, description="Punkte", layout=widgets.Layout(width="520px"))
pts_in  = widgets.Checkbox(value=True, description="Punkte markieren")
grid_in = widgets.Checkbox(value=True, description="Gitter")

controls_row1 = widgets.HBox([expr_in])
controls_row2 = widgets.HBox([xmin_in, xmax_in, pts_in, grid_in])
controls_row3 = widgets.HBox([n_in])

ui_box = widgets.VBox([controls_row1, controls_row2, controls_row3, out_panel])

# live updates
def _on_change(_):
    run_kurvendiskussion(expr_in.value, xmin_in.value, xmax_in.value, n_in.value, pts_in.value, grid_in.value)

for w in [expr_in, xmin_in, xmax_in, n_in, pts_in, grid_in]:
    w.observe(_on_change, names="value")

# -----------------------------
# Beschreibung als Accordion (im gleichen weißen Kasten)
# -----------------------------
info_out = widgets.Output()
with info_out:
    display(Markdown(r"""
<div style="padding: 15px;">

Dieses Tool ermöglicht es, eine **beliebige reelle Funktion** einzugeben und automatisch eine Kurvendiskussion
inklusive **grafischer Darstellung** zu erhalten.

**Eingabe (SymPy-Syntax):**
- Potenzen: `**`  (z.B. `x**2`)
- Wurzel: `sqrt(...)`
- Logarithmus: `ln(...)` oder `log(...)`
- Kreiszahl: `pi`
- Betrag: `abs(x)` oder `Abs(x)`

**Beispiele:**
- `x**3 - 2*x`
- `ln(1+x)`
- `x*sqrt(16-x**2)`
- `(x**2-1)/(x-1)`
- `sin(x)/x`

**Regler:**
- **xmin/xmax:** sichtbarer Bereich der x-Achse (Zoom, Asymptoten sichtbar machen)
- **Punkte:** Glätte vs. Performance
- **Punkte markieren:** Nullstellen / kritische Punkte / Wendepunkt-Kandidaten (soweit gefunden)
- **Gitter:** Raster ein/aus

**Automatisch berechnet (soweit möglich):**
- $f'(x)$, $f''(x)$
- Definitionsbereich (reell & stetig, soweit bestimmbar)
- Nullstellen / kritische Stellen / Wendestellen-Kandidaten
- Asymptoten (heuristisch: vertikal/horizontal/schief)

**Hinweis:** Bei komplizierten Funktionen kann SymPy einzelne Mengen/Nullstellen nicht geschlossen lösen.
Dann siehst du statt einer Liste ggf. ein Set wie `ConditionSet(...)`.

</div>
"""))

hint_out = widgets.Output()
with hint_out:
    display(Markdown(r"""
<div style="padding: 15px;">

### ℹ️ Praktische Hinweise

- Wenn Widgets in VS Code nicht angezeigt werden:  
  Stelle sicher, dass die **Jupyter Extension** aktiv ist und `ipywidgets` installiert ist.
- Wenn ein Plot Lücken hat: Das ist oft **gewollt** (z.B. Definitionslücken, Polstellen).
- Für Logarithmen: `ln(1+x)` ist nur für **x > -1** reell.

</div>
"""))

acc = widgets.Accordion(children=[info_out, hint_out])
acc.set_title(0, "📖 Anleitung anzeigen")
acc.set_title(1, "ℹ️ Hinweise anzeigen")

# -----------------------------
# Alles in EINEM weißen Kasten ausgeben
# -----------------------------
with main_out:
    display(Markdown(r"""
<div style="
background-color: white;
padding: 25px;
border-radius: 16px;
box-shadow: 0 4px 16px rgba(0,0,0,0.08);
margin: 20px 0;
">

### 📊 Interaktives Kurvendiskussions-Tool
<a id="plotkd"></a>

</div>
"""))
    display(acc)
    display(ui_box)

display(main_out)

# initial run
run_kurvendiskussion(expr_in.value, xmin_in.value, xmax_in.value, n_in.value, pts_in.value, grid_in.value)
```

```python
from IPython.display import display, Markdown
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets

# ============================
# EIN gemeinsamer Output
# ============================
main_out = widgets.Output()

# ============================
# Tipp
# ============================
tip_out = widgets.Output()
with tip_out:
    display(Markdown(r"""
<div style="padding: 15px;">

### 💡 Tipp (Asymptoten)

Für eine Funktion $f(x)$ prüfst du typischerweise drei Arten:

**1) Senkrechte Asymptoten (Polstellen)**
- Suche Stellen, wo der Nenner $=0$ ist (bei Brüchen).
- Prüfe: Wird der Zähler dort **nicht** auch $0$?  
  → Dann ist es wirklich eine Polstelle (vertikale Asymptote).

**2) Waagerechte Asymptoten**
- Berechne Grenzwerte:
$$
\lim_{x\to\infty} f(x), \qquad \lim_{x\to-\infty} f(x)
$$
- Wenn der Grenzwert endlich ist $=L$, dann ist $y=L$ eine waagerechte Asymptote.

**3) Schräge Asymptoten**
- Prüfe, ob $f(x)$ “wie eine Gerade” wächst.
- Bei rationalen Funktionen: **Polynomdivision** (Grad Zähler = Grad Nenner + 1 → schräg).
- Wenn nach Polynomdivision ein Restterm bleibt, der gegen $0$ geht → die Gerade ist die schräge Asymptote.

</div>
"""))

# ============================
# Lösung + Plots (nur sichtbar beim Aufklappen)
# ============================
sol_out = widgets.Output()

def plot_1():
    def f(x): return x / (4 - x**2)

    x1 = np.linspace(-5, -2.1, 400)
    x2 = np.linspace(-1.9, 1.9, 400)
    x3 = np.linspace(2.1, 5, 400)

    plt.figure()
    plt.plot(x1, f(x1))
    plt.plot(x2, f(x2))
    plt.plot(x3, f(x3))
    plt.axvline(-2, linestyle='--')
    plt.axvline(2, linestyle='--')
    plt.axhline(0, linestyle='--')
    plt.title(r"1) $f(x)=\frac{x}{4-x^2}$")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.ylim(-10, 10)
    plt.grid(True)
    plt.show()

def plot_2():
    def f(x):
        y = (x**2 - 3*x) / (2*x - 2)
        y[np.isclose(x, 1)] = np.nan   # verhindert Division durch 0
        return y

    def asym_schraeg(x):
        return 0.5*x - 1

    x = np.linspace(-5, 5, 2000)

    y = f(x)

    plt.figure()
    plt.plot(x, y)
    plt.axvline(1, linestyle="--")
    
    xx = np.linspace(-5, 5, 400)
    plt.plot(xx, asym_schraeg(xx), linestyle="--")

    plt.title(r"2) $h(x)=\frac{x^2-3x}{2x-2}$")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.ylim(-10, 10)
    plt.grid(True)
    plt.show()

def plot_3():
    def f(x): return x**2 * np.exp(-x)

    x = np.linspace(-5, 10, 1000)

    plt.figure()
    plt.plot(x, f(x), label=r"$g(x)=x^2 e^{-x}$")
    plt.axhline(0, linestyle="--", label=r"$y=0$")
    plt.title(r"3) $g(x)=x^2 e^{-x}$")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.ylim(-5, 50)
    plt.grid(True)
    plt.legend()
    plt.show()

with sol_out:
    display(Markdown(r"""
<div style="padding: 15px;">

## 📘 Lösung Aufgabe 4.8
<a id="aufgabe48-loesung"></a>

---

### 1)  $f(x) = \frac{x}{4 - x^2}$

**Senkrechte Asymptoten:**
$$
4 - x^2 = 0 \Rightarrow x = \pm 2
$$
Da $x\neq 0$ an $\pm 2$, sind das echte Polstellen:
$$
\boxed{x=-2 \ \text{und}\ x=2}
$$

**Waagerechte Asymptote:**
Für $x\to \pm\infty$:
$$
f(x)\sim \frac{x}{-x^2}=-\frac{1}{x}\to 0
$$
$$
\boxed{y=0}
$$

---

### 2)  $h(x) = \frac{x^2 - 3x}{2x - 2}$

**Senkrechte Asymptote:**
$$
2x-2=0 \Rightarrow x=1
$$
Zähler bei $x=1$ ist $1-3=-2\neq 0$ ⇒ Polstelle:
$$
\boxed{x=1}
$$

**Schräge Asymptote (Polynomdivision):**
$$
h(x)=\frac{1}{2}x-1-\frac{2}{2x-2}
$$
Der Restterm $\to 0$ für $x\to\pm\infty$, also:
$$
\boxed{y=\frac{1}{2}x-1}
$$

---

### 3)  $g(x)=x^2e^{-x}$

**Senkrechte Asymptoten:**  
Keine (kein Nenner, der $0$ werden kann).

**Waagerechte Asymptote:**
$$
\lim_{x\to+\infty}x^2e^{-x}
=
\lim_{x\to+\infty}\frac{x^2}{e^x}
=0
$$
$$
\boxed{y=0 \ \text{für}\ x\to +\infty}
$$

Für $x\to-\infty$ gilt $e^{-x}=e^{|x|}\to\infty$ und $x^2\to\infty$,
also $x^2e^{-x}\to\infty$ ⇒ keine Asymptote dort.

**Schräge Asymptote:**  
Keine, da der Grenzwert für $x\to +\infty$ bereits $0$ ist (waagerecht).

---

## 📉 Plots / grafische Kontrolle

</div>
"""))

    # Plots nur sichtbar im aufgeklappten Lösungsbereich
    plot_1()
    plot_2()
    plot_3()

# ============================
# Accordion (im weißen Kasten)
# ============================
acc = widgets.Accordion(children=[tip_out, sol_out])
acc.set_title(0, "💡 Tipp anzeigen")
acc.set_title(1, "📘 Lösung anzeigen (inkl. Plots)")

# ============================
# Alles in EINEM weißen Kasten
# ============================
with main_out:
    display(Markdown(r"""
<div style="
background-color: white;
padding: 25px;
border-radius: 16px;
box-shadow: 0 4px 16px rgba(0,0,0,0.08);
margin: 20px 0;
">

<a id="aufgabe8"></a>
# 📝 Aufgabe 4.8  
**Titel:** Asymptoten

Man bestimme die (waagerechten bzw. senkrechten bzw. schrägen Asymptoten der folgenden Funktionen)

### 1) $f(x) = \frac{x}{4 - x^2}$
### 2) $h(x) = \frac{x^2 - 3x}{2x - 2}$
### 3) $g(x) = x^2 e^{-x}$

</div>
"""))
    display(acc)

display(main_out)
```

```python
# ==========================================
# 📊 Interaktives Asymptoten-Tool (NEU & STABIL)
# ✅ funktioniert in Jupyter / VS Code Notebook
# ✅ eine einzige Ausgabe
# ✅ Button + Enter + Live-Update
# ✅ vertikale / horizontale / schräge Asymptoten
# ✅ hebbare Lücken optional
# ==========================================

import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
import ipywidgets as widgets

from IPython.display import display, Markdown, clear_output
from sympy.parsing.sympy_parser import (
    parse_expr,
    standard_transformations,
    implicit_multiplication_application,
    convert_xor
)

# -------------------------------------------------
# 1) SymPy-Grundlagen
# -------------------------------------------------
x = sp.Symbol("x", real=True)

TRANSFORMATIONS = standard_transformations + (
    implicit_multiplication_application,
    convert_xor,   # erlaubt z.B. x^2 -> x**2
)

LOCAL_DICT = {
    "x": x,
    "pi": sp.pi,
    "E": sp.E,

    "sin": sp.sin,
    "cos": sp.cos,
    "tan": sp.tan,
    "cot": sp.cot,

    "asin": sp.asin,
    "acos": sp.acos,
    "atan": sp.atan,

    "sinh": sp.sinh,
    "cosh": sp.cosh,
    "tanh": sp.tanh,

    "exp": sp.exp,
    "log": sp.log,
    "ln": sp.log,

    "sqrt": sp.sqrt,
    "abs": sp.Abs,
    "Abs": sp.Abs,
    "sign": sp.sign,
}

def parse_function(expr_str: str) -> sp.Expr:
    expr_str = expr_str.strip()
    if not expr_str:
        raise ValueError("Bitte eine Funktion eingeben.")
    expr = parse_expr(
        expr_str,
        local_dict=LOCAL_DICT,
        transformations=TRANSFORMATIONS,
        evaluate=True
    )
    return sp.simplify(expr)

def latex_of(expr):
    try:
        return sp.latex(expr)
    except Exception:
        return str(expr)

def real_unique(values, tol=1e-8):
    vals = sorted(values)
    out = []
    for v in vals:
        if not out or abs(v - out[-1]) > tol:
            out.append(v)
    return out

def side_to_text(side):
    return "+\\infty" if side == sp.oo else "-\\infty"

# -------------------------------------------------
# 2) Asymptoten-Analyse
# -------------------------------------------------
def get_denominator_candidates(expr: sp.Expr):
    """
    Kandidaten für Polstellen:
    reelle Nullstellen des Nenners von together(expr)
    """
    expr_t = sp.together(expr)
    num, den = sp.fraction(expr_t)
    den = sp.factor(den)

    if den == 1:
        return []

    candidates = []

    try:
        sols = sp.solve(sp.Eq(den, 0), x)
        for s in sols:
            s_num = sp.N(sp.simplify(s))
            if getattr(s_num, "is_real", False):
                candidates.append(float(s_num))
    except Exception:
        pass

    # Falls solve nichts liefert, notfalls numerisch bei Polynomen
    try:
        poly = sp.Poly(den, x)
        for r in poly.nroots():
            if abs(sp.im(r)) < 1e-10:
                candidates.append(float(sp.re(r)))
    except Exception:
        pass

    return real_unique(candidates)

def classify_vertical_asymptotes(expr: sp.Expr):
    """
    Rückgabe:
    vertical   = [(a, L_minus, L_plus), ...]
    removable  = [(a, y), ...]
    undecided  = [a, ...]
    """
    candidates = get_denominator_candidates(expr)

    vertical = []
    removable = []
    undecided = []

    for a in candidates:
        try:
            Lm = sp.limit(expr, x, a, dir="-")
            Lp = sp.limit(expr, x, a, dir="+")

            inf_left = (Lm is sp.oo) or (Lm is -sp.oo)
            inf_right = (Lp is sp.oo) or (Lp is -sp.oo)

            if inf_left or inf_right:
                vertical.append((a, Lm, Lp))
            else:
                # hebbare Lücke?
                try:
                    L = sp.limit(expr, x, a)
                    if getattr(L, "is_real", False) or getattr(L, "is_finite", False):
                        removable.append((a, sp.simplify(L)))
                    else:
                        undecided.append(a)
                except Exception:
                    undecided.append(a)

        except Exception:
            undecided.append(a)

    return vertical, removable, undecided

def horizontal_asymptotes(expr: sp.Expr):
    """
    y = L für x -> ±∞
    """
    result = []

    for side in (sp.oo, -sp.oo):
        try:
            L = sp.limit(expr, x, side)
            if getattr(L, "is_finite", False) is True:
                result.append((side, sp.simplify(L)))
        except Exception:
            pass

    return result

def oblique_asymptotes(expr: sp.Expr):
    """
    Schräge Asymptoten y = m*x + b
    nur wenn m endlich und != 0, b endlich
    """
    result = []

    for side in (sp.oo, -sp.oo):
        try:
            m = sp.limit(expr / x, x, side)
            if getattr(m, "is_finite", False) is True and sp.simplify(m) != 0:
                b = sp.limit(expr - m * x, x, side)
                if getattr(b, "is_finite", False) is True:
                    result.append((side, sp.simplify(m), sp.simplify(b)))
        except Exception:
            pass

    return result

# -------------------------------------------------
# 3) Numerische Plot-Helfer
# -------------------------------------------------
def safe_lambdify(expr: sp.Expr):
    return sp.lambdify(x, expr, modules=["numpy"])

def evaluate_real(f_np, xs):
    with np.errstate(all="ignore"):
        ys = f_np(xs)

    ys = np.asarray(ys)

    # komplexe Werte aussortieren
    if np.iscomplexobj(ys):
        ys = np.where(np.abs(np.imag(ys)) < 1e-9, np.real(ys), np.nan)

    ys = np.array(ys, dtype=float)
    ys[~np.isfinite(ys)] = np.nan
    return ys

def build_segments(x_min, x_max, vertical_asymptotes, margin_ratio=1e-4):
    """
    Zerschneidet den Bereich an vertikalen Asymptoten,
    damit nicht über Polstellen hinweg verbunden wird.
    """
    cuts = [x_min, x_max]
    for a, _, _ in vertical_asymptotes:
        if x_min < a < x_max:
            cuts.append(a)

    cuts = sorted(set(cuts))
    width = max(1.0, abs(x_max - x_min))
    eps = margin_ratio * width

    segments = []
    for i in range(len(cuts) - 1):
        left = cuts[i]
        right = cuts[i + 1]

        if any(abs(left - a) < 1e-12 for a, _, _ in vertical_asymptotes):
            left = left + eps
        if any(abs(right - a) < 1e-12 for a, _, _ in vertical_asymptotes):
            right = right - eps

        if right > left:
            segments.append((left, right))

    return segments

# -------------------------------------------------
# 4) Widgets
# -------------------------------------------------
func_input = widgets.Text(
    value="(x**2-1)/(x-1)",
    description="f(x) =",
    layout=widgets.Layout(width="700px")
)

x_min_w = widgets.FloatText(value=-10, description="x_min", layout=widgets.Layout(width="170px"))
x_max_w = widgets.FloatText(value=10, description="x_max", layout=widgets.Layout(width="170px"))
y_min_w = widgets.FloatText(value=-10, description="y_min", layout=widgets.Layout(width="170px"))
y_max_w = widgets.FloatText(value=10, description="y_max", layout=widgets.Layout(width="170px"))

n_points_w = widgets.IntSlider(
    value=1000, min=200, max=4000, step=100,
    description="Punkte",
    layout=widgets.Layout(width="500px")
)

show_removable_w = widgets.Checkbox(
    value=True,
    description="Hebbare Lücken markieren"
)

live_update_w = widgets.Checkbox(
    value=True,
    description="Live-Update"
)

update_button = widgets.Button(
    description="Plot aktualisieren",
    button_style="primary",
    icon="refresh"
)

output = widgets.Output()

# -------------------------------------------------
# 5) Render-Funktion
# -------------------------------------------------
def render(_=None):
    with output:
        clear_output(wait=True)

        # -------------------------
        # Eingaben lesen
        # -------------------------
        expr_str = func_input.value.strip()

        try:
            x_min = float(x_min_w.value)
            x_max = float(x_max_w.value)
            y_min = float(y_min_w.value)
            y_max = float(y_max_w.value)
            n_points = int(n_points_w.value)
        except Exception:
            display(Markdown(
                """
<div style="background:#fff3f3; padding:14px; border-radius:12px; border:1px solid #ffd1d1;">
<b>❌ Fehler:</b> Achsenwerte oder Punktzahl sind ungültig.
</div>
                """
            ))
            return

        if x_max <= x_min:
            display(Markdown(
                """
<div style="background:#fff3f3; padding:14px; border-radius:12px; border:1px solid #ffd1d1;">
<b>❌ Fehler:</b> Es muss <code>x_min &lt; x_max</code> gelten.
</div>
                """
            ))
            return

        if y_max <= y_min:
            display(Markdown(
                """
<div style="background:#fff3f3; padding:14px; border-radius:12px; border:1px solid #ffd1d1;">
<b>❌ Fehler:</b> Es muss <code>y_min &lt; y_max</code> gelten.
</div>
                """
            ))
            return

        # -------------------------
        # Funktion parsen
        # -------------------------
        try:
            expr = parse_function(expr_str)
        except Exception as e:
            display(Markdown(f"""
<div style="background:#fff3f3; padding:14px; border-radius:12px; border:1px solid #ffd1d1;">
<b>❌ Konnte die Funktion nicht lesen.</b><br><br>
<b>Fehler:</b> <code>{str(e)}</code><br><br>
<b>Beispiele:</b>
<ul>
<li><code>1/(x-2)</code></li>
<li><code>(x**2-1)/(x-1)</code></li>
<li><code>x + 1/x</code></li>
<li><code>sin(x)/x</code></li>
<li><code>(2*x**2+3)/(5*x**2+7)</code></li>
</ul>
</div>
            """))
            return

        # -------------------------
        # Analyse
        # -------------------------
        vertical, removable, undecided = classify_vertical_asymptotes(expr)
        horizontal = horizontal_asymptotes(expr)
        oblique = oblique_asymptotes(expr)

        # wenn schräge Asymptote existiert, horizontale i.d.R. nicht extra anzeigen
        # (rein optisch oft sinnvoll)
        # symbolisch lassen wir aber beides zu

        # -------------------------
        # Ausgabe oben
        # -------------------------
        vert_text = "keine"
        if vertical:
            vert_text = ", ".join([f"$x = {a:.6g}$" for a, _, _ in vertical])

        horiz_text = "keine"
        if horizontal:
            horiz_text = "; ".join([
                f"$x \\to {side_to_text(side)}: \\; y = {latex_of(L)}$"
                for side, L in horizontal
            ])

        oblique_text = "keine"
        if oblique:
            oblique_text = "; ".join([
                f"$x \\to {side_to_text(side)}: \\; y = {latex_of(m)}x + ({latex_of(b)})$"
                for side, m, b in oblique
            ])

        display(Markdown(rf"""
<div style="background:white; padding:18px; border-radius:16px; box-shadow:0 4px 16px rgba(0,0,0,0.06); margin:10px 0 16px 0;">
<h3>📌 Ergebnis</h3>

**Eingegebene Funktion:**

$$
f(x) = {latex_of(expr)}
$$

- 🔴 **Vertikale Asymptoten:** {vert_text}  
- 🟢 **Horizontale Asymptoten:** {horiz_text}  
- 🟠 **Schräge Asymptoten:** {oblique_text}
</div>
        """))

        # -------------------------
        # Plot
        # -------------------------
        try:
            f_np = safe_lambdify(expr)
        except Exception as e:
            display(Markdown(f"""
<div style="background:#fff3f3; padding:14px; border-radius:12px; border:1px solid #ffd1d1;">
<b>❌ Die Funktion konnte nicht numerisch geplottet werden.</b><br>
<code>{str(e)}</code>
</div>
            """))
            return

        fig, ax = plt.subplots(figsize=(10, 5.5))

        # Funktion segmentweise
        segments = build_segments(x_min, x_max, vertical)

        first_fun = True
        for left, right in segments:
            xs = np.linspace(left, right, n_points)
            ys = evaluate_real(f_np, xs)
            ax.plot(xs, ys, linewidth=2, label="f(x)" if first_fun else None)
            first_fun = False

        # Achsen
        ax.axhline(0, linewidth=1, color="black")
        ax.axvline(0, linewidth=1, color="black")

        # vertikale Asymptoten
        first_v = True
        for a, _, _ in vertical:
            if x_min <= a <= x_max:
                ax.axvline(
                    a,
                    linestyle="--",
                    linewidth=2,
                    color="tab:red",
                    label="vertikale Asymptote" if first_v else None
                )
                first_v = False

        # horizontale Asymptoten
        first_h = True
        for side, L in horizontal:
            try:
                y0 = float(sp.N(L))
                if y_min <= y0 <= y_max:
                    ax.axhline(
                        y0,
                        linestyle="--",
                        linewidth=2,
                        color="tab:green",
                        label="horizontale Asymptote" if first_h else None
                    )
                    first_h = False
            except Exception:
                pass

        # schräge Asymptoten
        first_o = True
        xs_line = np.linspace(x_min, x_max, 800)
        for side, m, b in oblique:
            try:
                mf = float(sp.N(m))
                bf = float(sp.N(b))
                ys_line = mf * xs_line + bf
                ax.plot(
                    xs_line,
                    ys_line,
                    linestyle="--",
                    linewidth=2,
                    color="tab:orange",
                    label="schräge Asymptote" if first_o else None
                )
                first_o = False
            except Exception:
                pass

        # hebbare Lücken
        if show_removable_w.value and removable:
            first_r = True
            for a, y in removable:
                try:
                    af = float(a)
                    yf = float(sp.N(y))
                    if x_min <= af <= x_max and y_min <= yf <= y_max:
                        ax.plot(
                            [af], [yf],
                            marker="o",
                            markersize=8,
                            markerfacecolor="white",
                            markeredgecolor="black",
                            markeredgewidth=2,
                            linestyle="None",
                            label="hebbare Lücke" if first_r else None
                        )
                        first_r = False
                except Exception:
                    pass

        ax.set_xlim(x_min, x_max)
        ax.set_ylim(y_min, y_max)
        ax.set_xlabel("x")
        ax.set_ylabel("y")
        ax.set_title("Asymptoten-Tool")
        ax.grid(True, alpha=0.3)
        ax.legend(loc="best")

        plt.show()
        plt.close(fig)

        # Zusatzhinweise
        if removable:
            rem_text = ", ".join([f"$({a:.6g}, {latex_of(y)})$" for a, y in removable])
            display(Markdown(rf"""
<div style="background:#f8fbff; padding:14px; border-radius:12px; border:1px solid #d6ebff; margin-top:10px;">
<b>ℹ️ Hebbare Lücken:</b> {rem_text}
</div>
            """))

        if undecided:
            undec_text = ", ".join([f"$x = {a:.6g}$" for a in undecided])
            display(Markdown(rf"""
<div style="background:#fffdf2; padding:14px; border-radius:12px; border:1px solid #ffe8a6; margin-top:10px;">
<b>⚠️ Hinweis:</b> Bei folgenden Stellen konnte die Symbolik nicht alles eindeutig klassifizieren: {undec_text}
</div>
            """))

# -------------------------------------------------
# 6) Events
# -------------------------------------------------
def maybe_render(change=None):
    if live_update_w.value:
        render()

update_button.on_click(render)

# Enter / Änderung im Textfeld
func_input.observe(maybe_render, names="value")

# Bereichsänderungen
for w in [x_min_w, x_max_w, y_min_w, y_max_w, n_points_w, show_removable_w]:
    w.observe(maybe_render, names="value")

# -------------------------------------------------
# 7) Layout
# -------------------------------------------------
info_html = widgets.HTML(
    value="""
<div style="
    background:white;
    padding:25px;
    border-radius:16px;
    box-shadow:0 4px 16px rgba(0,0,0,0.08);
    margin:10px 0;
">
    <h2 style="margin-top:0;">📊 Interaktives Asymptoten-Tool</h2>

    <p>Gib eine Funktion in <b>SymPy-/Python-Syntax</b> ein, zum Beispiel:</p>
    <ul>
        <li><code>1/(x-2)</code></li>
        <li><code>(x**2-1)/(x-1)</code></li>
        <li><code>x + 1/x</code></li>
        <li><code>sin(x)/x</code></li>
        <li><code>(2*x**2+3)/(5*x**2+7)</code></li>
    </ul>

    <p><b>Erlaubt:</b> <code>sin</code>, <code>cos</code>, <code>tan</code>, <code>exp</code>, <code>log</code>, <code>ln</code>, <code>sqrt</code>, <code>Abs</code></p>

    <p><b>Farben im Plot:</b><br>
    🔵 Funktion &nbsp;&nbsp; 🔴 vertikale Asymptoten &nbsp;&nbsp; 🟢 horizontale Asymptoten &nbsp;&nbsp; 🟠 schräge Asymptoten</p>
</div>
"""
)

controls = widgets.VBox([
    func_input,
    widgets.HBox([x_min_w, x_max_w, y_min_w, y_max_w]),
    n_points_w,
    widgets.HBox([show_removable_w, live_update_w]),
    update_button
])

container = widgets.VBox([
    info_html,
    controls,
    output
])

display(container)

# Initiale Ausgabe
render()
```
